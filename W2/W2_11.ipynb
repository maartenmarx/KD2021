#### Familierelatie niet te definiren in predicaatlogica

Een stiefvader of stiefmoeder is niet goed te definiren in predicaatlogica. Dit komt omdat de stiefmoeder of -vader niet de biologische ouder is. De twee originele biologische ouders zijn namelijk de enige twee ouders van een kind, waar in deze opdracht de R(x,y) formule voor gebruikt wordt. Deze situatie komt men natuurlijk in het echte leven wel regelmatig tegen binnen families.

Geadopteerde kinderen. Tussen geadopteerde kinderen en ouders of andere personen in de familie is namelijk geen biologische relatie. Hierdoor is deze niet goed te definiren in predicaatlogica. Maar een geadopteerd kind ziet zijn niet biologische ouders in het echt vaak wel als vader en moeder.
#### Aantal beperkingen

Je bent alleen de broer of zus van je broer of zus

Je bent alleen de vader of moeder van je kinderen

Je bent alleen opa of oma van je kleinkinderen 

Je bent alleen de zoon of dochter van je ouders

## Uitwerking in Markdown
#### LHBTQ+ relaties
| Relatie | Formule | Uitleg |
| ------- | ------- | ------ |
| Hetero |  $ \forall x (Hetero(x)) \leftrightarrow R(x,y) \land ((M(x) \land V(y)) \vee (M(y) \land V(x))) $| x is alleen en slechts dan alleen hetero als x seskueel wordt aangetrokken tot y en x een man en y een vrouw is|of x een vrouw en y een man is. |
| Lesbienne |  $ \forall x (Lesbienne(x)) \leftrightarrow R(x, y) \land (V(y) \land V(x)) $ | x is alleen en slechts dan alleen een lesbienne als x seksueel wordt aangetrokken tot y en zowel x als y een vrouw zijn. |
| Homofiel |   $ \forall x (Homofiel(x)) \leftrightarrow R(x, y) \land (M(y) \land M(x)) $| x is alleen en slechts dan alleen een Homofiel als x seksueel wordt aangetrokken tot y en zowel x als y een man zijn. |
| Biseksueel | $ \forall x (Lesbienne(x)) \leftrightarrow R(x, y) \land (V(y) \vee M(y)) $| x is alleen en slechts dan alleen biseksueel als x seksueel wordt aangetrokken tot y en y een man of een vrouw is |
| Transgenders | Niet mogelijk | Hiervoor is er geen logische formule voor de relatie. Aan een relatie is namelijk niet te zien of iemand transgender is omdat een transgender bijvoorbeeld een homo kan zijn maar ook een lesbie. |
| Asexuelen | $ \forall x (Asexueel(x) \leftrightarrow \neg R(x, y)) $| x is alleen en slechts dan alleen asexueel als x niet seksueel wordt aangetrokken tot y. |
| Queer | Niet mogelijk | Een queer heeft geen duidelijke relatie omdat het geslacht van een queer niet duidelijk is en bijvoorbeeld ook kan verschillen per dag zoals bij iemand die genderflude is. Deze mensen vallen namelijk ook onder queers.
| Intersexueel | Niet mogelijk | Mensen die intersexueel zijn hebben eigenschappen van zowel mannen en vrouwen. Hierdoor kan er geen duidelijke logische relatie gemaakt worden voor intersexuelen omdat hun geslacht niet duidelijk is. 
| Onbenoemde | $ \forall x (Onbenoemde(x)) \leftrightarrow \neg(\exists(Lesbienne)  \vee  \exists(Homofiel) \vee \exists(Biseksueel) \vee \exists(Transgenders) \vee \exists(Asexuelen) \vee \exists(Queer) \vee \exists(Intersexueel) \vee \exists(Hetero)) $ | x is alleen en slechts dan alleen een onbenoemde als x niet lesbienne, homofiel, biseksueel, transgender, asexueel,  queer, intersexueel of hetero is. |



```python
relations = nx.DiGraph()

M = {"Homo", "Bi", "Transgender"}
V = {"Lesbienne", "Asexueel"}
X = {"Intersexueel", "Queer"}



relations.add_nodes_from(M)
relations.add_nodes_from(V)
relations.add_nodes_from(X)

relations.add_edges_from([("Homo", "Bi"), ("Bi", "Homo"), ("Homo", "Transgender"), ("Homo", "Queer"), ("Homo", "Intersexueel"), ("Transgender", "Homo"), ("Transgender", "Bi"), ("Bi", "Transgender"), ("Bi", "Lesbienne"), ("Bi", "Intersexueel"), ("Bi", "Asexueel"), ("Transgender", "Intersexueel"), ("Lesbienne", "Asexueel"), ("Lesbienne", "Queer"), ("Lesbienne", "Intersexueel"), ("Queer", "Lesbienne"), ("Intersexueel", "Queer"), ("Queer", "Intersexueel"), ("Queer", "Homo"), ("Queer", "Bi"), ("Queer", "Transgender"), ("Queer", "Lesbienne"), ("Queer", "Asexueel"), ("Intersexueel", "Homo"), ("Intersexueel", "Bi"), ("Intersexueel", "Transgender"), ("Intersexueel", "Lesbienne"), ("Intersexueel", "Asexueel"), ("Transgender", "Queer"), ("Transgender", "Intersexueel")])

for n in M:
    relations.nodes[n]['M']=True
    relations.nodes[n]['V']=False
for n in V:
    relations.nodes[n]['V']=True
    relations.nodes[n]['M']=False
for n in X:
    relations.nodes[n]['V']=False
    relations.nodes[n]['M']=False
    
def M(x, network=relations):
    return network.nodes[x]['M']
def V(x, network=relations):
    return network.nodes[x]['V']
def X(x, network=relations):
    return network.nodes[x]['X']
def R(x,y,network=relations):
    return network.has_edge(x,y)

Hetero = {(x,y) for x in relations for y in relations if R(x,y) and (M(x) and V(y) or V(x) and M(y))}
Lesbienne = {(x,y) for x in relations for y in relations if R(x,y) and (V(x) and V(y))}
Homofiel = {(x,y) for x in relations for y in relations if R(x,y) and (M(x) and M(y))}
Bisexueel = {(x,y) for x in relations for y in relations if R(x,y) and (V(y) or M(y))}
Asexueel = {(x,y) for x in relations for y in relations if not R(x,y)}
Onbenoemde = {(x,y) for x in relations if not (Hetero or lesbienne or Homofiel or Bisexueel or Asexueel)}

nx.draw(relations, with_labels=True)
```

```python
print("\nAlle Hetero relaties\n")
for person in Hetero:
    print(person[0], "heeft een relatie met", person[1])

print("\nAlle Lesbienne relaties\n")
for person in Lesbienne:
    print(person[0], "heeft een relatie met", person[1])
    
print("\nAlle Homofiele relaties\n")
for person in Homofiel:
    print(person[0], "heeft een relatie met", person[1])
    
print("\nAlle Bisexuele relaties\n")
for person in Bisexueel:
    print(person[0], "heeft een relatie met", person[1])
    
print("\nAlle Asexuele relaties\n")
for person in Asexueel:
    print(person[0], "heeft een relatie met", person[1])

print("\nAlle Onbenoemde relaties\n")
for person in Onbenoemde:
    print(person[0], "heeft een relatie met", person[1])
```
