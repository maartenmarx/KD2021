## 1 De gebruikelijke familierelaties

* We werken met predicaatlogische modellen, en met een basistaal met 2 eigenschappen: $M$ en $V$, die staan voor de 2 traditionele geslachten, en 1 twee plaatsige relatie $R$, die staat voor *is ouder van*. Hieronder staat dus dat *piet een man is, en joke een vrouw en dat piet de ouder is van joke*:

$$M(piet), V(joke), R(piet,joke).$$
* Wij gaan nu 2 dingen doen:
    1. beperkingen opleggen aan de relaties
    2. nieuwe relaties definieren in termen van de 3 primtieve relaties $M,V,R$
        * in die nieuwe relaties kan je natuurlijk andere eerder gedefinieerde relaties weer gebruiken.
* **Voorbeelden**
* *Beperking:* 
    * Iedereen is man of vrouw (en nooit allebei tegelijk).
    * Iedereen heeft exact 2 ouders, een man en een vrouw.
* *Nieuwe relatie*
    * $Vader(x,y)$ precies dan en slechts dan alleen als x een ouder is van y en x een man is.
        * $\forall xy( Vader(x,y) \leftrightarrow R(x,y) \wedge M(x))$
    * $Opa(x,y)$ hiervoor gebruik je natuurlijk het $\exists z$    connectief.
        * Je kunt dit netjes definieren in termen van ouder en vader.

```python
import networkx as nx
%matplotlib inline
```

```python
# Voorbeeld om je op weg te helpen

H= nx.house_graph()  # In het echt neem je natuurlijk een nx.Digraph !
H[1]
nx.draw_networkx(H)
M,V= {1,3,0},{2,4} # de mannen en de vrouwen in dit netwerk
# Nu even de attributen M en V goed zetten
for n in M:
    H.nodes[n]['M']=True
    H.nodes[n]['V']=False
for n in V:
    H.nodes[n]['V']=True
    H.nodes[n]['M']=False    
 

# wat handige afkortingen om de comprehensies zo veel mogelijk op je formules te laten lijken
G=H
def M(x, network=G):
    return network.nodes[x]['M']
def V(x, network=G):
    return network.nodes[x]['V']
def R(x,y,network=G):
    return network.has_edge(x,y)

# Vader(x,y) iff R(x,y) & M(x)
# Bijna hetzelfde als in deze comprehensie toch?
Vader= {(x,y) for x in H for y in H if R(x,y,H) and M(x,H)}
print(Vader)

```

```python
# Verdeling van verschillende geslachten (man, vrouw, niet geboren man, niet geboren vrouw, non-binari,interseksueel, overig)
# Geslachtsvoorkeur kunnen niet worden aangetoond in relaties, hierdoor hebben wij zelf de nodes gegroepeerd om dit aan te kunnen geven.
#  Het verschil tussen non-binair(NB) en overig(NVT) is dat non-binair geen geslacht heeft, en overig een geslacht dat geen man of vrouw is
M,V,NM,NV,NB,I,NVT= {'Hans', 'Bert', 'Sjon'},{'Lisa','Quincy'},{'Tim'},{'Inge'},{'Pam'},{'Sam'}, {'Peer'}
K= nx.DiGraph()  # In het echt neem je natuurlijk een nx.Digraph !
K.add_edges_from([('Sam', 'Pam'),('Pam','Sam'),('Bert', 'Hans'),('Hans', 'Bert'),('Lisa','Quincy'),('Quincy','Lisa'),('Tim','Inge'),('Bert','Lisa'),('Lisa','Bert'), ('Inge','Tim'),('Peer','Inge'),('Inge','Peer')])
K.add_node('Sjon')
# Check bij welke groep persoon hoort
for n in M:
    K.nodes[n]['M']=True
    K.nodes[n]['V']=False 
    K.nodes[n]['NM']=False 
    K.nodes[n]['NV']=False 
    K.nodes[n]['NB']=False
    K.nodes[n]['I']=False
    K.nodes[n]['NVT']=False
for n in V:
    K.nodes[n]['M']=False 
    K.nodes[n]['V']=True
    K.nodes[n]['NM']=False 
    K.nodes[n]['NV']=False 
    K.nodes[n]['NB']=False
    K.nodes[n]['I']=False
    K.nodes[n]['NVT']=False
    
for n in NM:
    K.nodes[n]['M']=False 
    K.nodes[n]['V']=False
    K.nodes[n]['NM']=True 
    K.nodes[n]['NV']=False 
    K.nodes[n]['NB']=False
    K.nodes[n]['I']=False
    K.nodes[n]['NVT']=False
    
for n in NV:
    K.nodes[n]['M']=False 
    K.nodes[n]['V']=False
    K.nodes[n]['NM']=False 
    K.nodes[n]['NV']=True 
    K.nodes[n]['NB']=False
    K.nodes[n]['I']=False
    K.nodes[n]['NVT']=False
for n in NB:
    K.nodes[n]['M']=False
    K.nodes[n]['V']=False 
    K.nodes[n]['NM']=False 
    K.nodes[n]['NV']=False 
    K.nodes[n]['NB']=True 
    K.nodes[n]['I']=False
    K.nodes[n]['NVT']=False
for n in I:
    K.nodes[n]['M']=False
    K.nodes[n]['V']=False 
    K.nodes[n]['NM']=False 
    K.nodes[n]['NV']=False 
    K.nodes[n]['NB']=False
    K.nodes[n]['I']=True
    K.nodes[n]['NVT']=False

for n in NVT:
    K.nodes[n]['M']=False
    K.nodes[n]['V']=False 
    K.nodes[n]['NM']=False 
    K.nodes[n]['NV']=False 
    K.nodes[n]['NB']=False
    K.nodes[n]['I']=False
    K.nodes[n]['NVT']=True    
    
def M(x, network=K):
    return network.nodes[x]['M']
def V(x, network=K):
    return network.nodes[x]['V']
def NM(x, network=K):
    return network.nodes[x]['NM']
def NV(x, network=K):
    return network.nodes[x]['NV']
def NB(x, network=K):
    return network.nodes[x]['NB']
def I(x, network=K):
    return network.nodes[x]['I']
def NVT(x, network=K):
    return network.nodes[x]['NVT']
def R(x,y,network=K):
    return network.has_edge(x,y)
# comprehensions voor checken seksuele/geslachts voorkeur

Homoseksueel= {(x) for x in K for y in K if not {z for z in K if R(x,z,K) and V(z) } and R(x,y,K) and M(x) and M(y)}

Lesbisch = {(x) for x in K for y in K if not {z for z in K if R(x,z,K) and M(z) } and R(x,y,K) and V(x) and V(y)}

Biseksueel= {(x) for x in K for y in K for z in K if R(x,y,K) and R(x,z,K) and M(y,K) and V(z,K)}

Transgender= {(x) for x in K if (NM(x,K) or NV(x,K)) and not (M(x,K) or V(x,K))}

Intersekse = {(x) for x in K if I(x,K)}

Nonbinair = {(x) for x in K if NB(x,K)}

Q = {(x) for x in K if x in Homoseksueel or Lesbisch or Biseksueel or Intersekse or Nonbinair}

plus = {(x) for x in K if NVT(x,K)}
# plot van netwerk
pos = nx.spring_layout(K, k=0.75, iterations=1)
nx.draw_networkx(K,pos)
```
