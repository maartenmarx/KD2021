Kinderen
* $Zoon(x,y)$ precies dan en slechts dan alleen als y ouder is van x en x een man is.
    * $ \forall xy( Zoon(x,y) \leftrightarrow R(y,x) \wedge M(x)) $


* $Dochter(x,y)$ precies dan en slechts dan alleen als y ouder is van x en x een vrouw is.
    * $ \forall xy( Dochter(x,y) \leftrightarrow R(y,x) \wedge V(x)) $
    
<h6>Rang 2</h6>

* $Broer(x,y)$ precies dan en slechts dan alleen als x een man is en z een ouder is van x, z een ouder is van y.
    * $ \forall xy( Broer(x,y) \leftrightarrow M(x) \wedge \exists z(R(z,x) \wedge R(z,y))) $


* $Zus(x,y)$ precies dan en slechts dan alleen als x een vrouw is en z een ouder is van x, z een ouder is van y.
    * $ \forall xy( Zus(x,y) \leftrightarrow V(x) \wedge \exists z(R(z,x) \wedge R(z,y))) $

Grootouders
* $Opa(x,y)$ precies dan en slechts dan alleen als x vader is van z en z een ouder is van y.
    * $ \forall xy( Opa(x,y) \leftrightarrow \exists z(Vader(x,z) \wedge R(z,y))) $


* $Oma(x,y)$ precies dan en slechts dan alleen als x moeder is van z en z een ouder is van y.
    * $ \forall xy( Oma(x,y) \leftrightarrow \exists z(Moeder(x,z) \wedge R(z,y))) $


Kleinkinderen
* $Kleinzoon(x,y)$ precies dan en slechts dan alleen als y oma of opa is van x en x een man is.
    * $ \forall xy( Kleinzoon(x,y) \leftrightarrow (Oma(y,x) \vee Opa(y,x)) \wedge M(x)) $


* $Kleindochter(x,y)$ precies dan en slechts dan alleen als y oma of opa is van x en x een vrouw is.
    * $ \forall xy( Kleindochter(x,y) \leftrightarrow (Oma(y,x) \vee Opa(y,x)) \wedge V(x)) $

Overig
* $Oom(x,y)$ precies dan en slechts dan alleen als x een broer is van z en z ouder is van y.
    * $ \forall xy( Oom(x,y) \leftrightarrow \exists z(Broer(x,z) \wedge R(z,y))) $


* $Tante(x,y)$ precies dan en slechts dan alleen als x een zus is van z en z ouder is van y.
    * $ \forall xy( Tante(x,y) \leftrightarrow \exists z(Zus(x,z) \wedge R(z,y))) $


* $Neef(x,y)$ precies dan en slechts dan alleen als y oom of tante is van x en x een man is.
    * $ \forall xy( Neef(x,y) \leftrightarrow (Oom(y,x) \vee Tante(y,x)) \wedge M(x)) $


* $Nicht(x,y)$ precies dan en slechts dan alleen als y oom of tante is van x en x een vrouw is.
    * $ \forall xy( Nicht(x,y) \leftrightarrow (Oom(y,x) \vee Tante(y,x)) \wedge V(x)) $
    
Beperkingen
* Iemand kan geen relatie met zichzelf hebben.
    * Voor een relatie zijn twee personen nodig.
* Iemand kan geen broer/zus zijn met zijn/haar ouder.
    * Een broer/zus relatie kan alleen bestaan tussen twee personen met dezelfde ouders.
* Iemand kan niet tegelijk een ouder en kind zijn van n persoon.
    * Ouder/kind relatie is asymmetrisch.
* Iemand kan niet zowel een man als een vrouw zijn.
    * Een persoon is of een man, of een vrouw.
* Een ouder kan alleen een vader/moeder relatie hebben tegenover zijn/haar kind (dus deze ouder kan niet tegelijk een oom/tante zijn voor deze persoon).
    * Tegenover een ander persoon is maar n rol tegelijk mogelijk.
    
Deze laatste is wellicht in het echt wel mogelijk (je oom of opa kan als een vader voor je zijn), maar strikt logisch gezien kan je oom niet tegelijk je vader zijn (of je tante kan bijvoorbeeld niet tegelijk je zus zijn). Hier is de uitzondering in het model ouder en vader/moeder, maar deze rollen zijn technisch gezien hetzelfde (met als enige verschil dat vader/moeder gecategoriseerd wordt op gender), dus dat maakt in dit geval geen verschil voor deze beperking.

* Primitieven:
    * $M$ - Bezit mannelijke geslachtsdelen.
    * $V$ - Bezit vrouwlijke geslachtsdelen.
    * $IM$ - Identificeert zich als man.
    * $IV$ - Identificeert zich als vrouw.
    * $R$ - Houdt van


* Relaties:
    * $CisM(x)$ - x is cis man precies dan en alleen dan als x mannelijke geslachtsdelen bezit en zich identificeert als man.
        * $ \forall x( CisM(x) \leftrightarrow M(x) \wedge IM(x)) $ 
        
    * $CisV(x)$ - x is cis vrouw precies dan en alleen dan als x vrouwelijke geslachtsdelen bezit en zich identificeert als vrouw.
        * $ \forall x( CisV(x) \leftrightarrow V(x) \wedge IV(x)) $
        
    * $Hetero(x)$ - x is hetero precies dan en alleen dan als x een cis man is en alleen van cis vrouwen houdt of als x een cis vrouw is en alleen van cis mannen houdt.
        * $ \forall x( Hetero(x) \leftrightarrow (CisM(x) \wedge \exists y(CisV(y) \wedge R(x,y)) \wedge \neg \exists y(\neg CisV(y) \wedge R(x,y))) \vee (CisV(x) \wedge \exists y(CisM(y) \wedge R(x,y)) \wedge \neg \exists y(\neg CisM(y) \wedge R(x,y)))  ) $
        
    * $L(x)$ - x is lesbisch precies dan en alleen dan als x een cis vrouw is en alleen van cis vrouwen houdt.
        * $ \forall x( L(x) \leftrightarrow CisV(x) \wedge \exists y(CisV(y) \wedge R(x,y)) \wedge \neg \exists y(\neg CisV(y) \wedge R(x,y,)) )  $
        
    * $H(x)$ - x is homo precies dan en alleen dan als x een cis man is en alleen van cis mannen houdt.
        * $ \forall x( H(x) \leftrightarrow CisM(x) \wedge \exists y(CisM(y) \wedge R(x,y)) \wedge \neg \exists y(\neg CisM(y) \wedge R(x,y,)) )  $
        
    * $B(x)$ - x is biseksueel precies dan en alleen dan als x alleen van cis mannen of cis vrouwen houdt.
        * $ \forall x( B(x) \leftrightarrow \exists y(CisM(y) \wedge R(x,y)) \wedge \exists y(CisV(y) \wedge R(x,y)) \wedge \neg \exists y(\neg (CisM(y) \vee CisV(y)) \wedge R(x,y)) ) $
        
    * $T(x)$ - x is transgender precies dan en alleen dan als x mannelijke geslachtsdelen, maar zich identificeert als vrouw of als x vrouwelijke geslachtsdelen bezit, maar zich identificeert als man.
        * $ \forall x( T(x) \leftrightarrow (M(x) \wedge IV(x)) \vee (V(x) \wedge IM(x)) ) $

        * $ \forall x( Q(x) \leftrightarrow \neg IM(x) \wedge \neg IV(x) ) $

        * $ \forall x( I(x) \leftrightarrow (\neg M(x) \wedge \neg V(x)) \vee (M(x) \wedge V(x)) ) $
