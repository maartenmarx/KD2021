## 1 De gebruikelijke familierelaties

* We werken met predicaatlogische modellen, en met een basistaal met 2 eigenschappen: $M$ en $V$, die staan voor de 2 traditionele geslachten, en 1 twee plaatsige relatie $R$, die staat voor *is ouder van*. Hieronder staat dus dat *piet een man is, en joke een vrouw en dat piet de ouder is van joke*:

$$M(piet), V(joke), R(piet,joke).$$
* Wij gaan nu 2 dingen doen:
    1. beperkingen opleggen aan de relaties
    2. nieuwe relaties definieren in termen van de 3 primtieve relaties $M,V,R$
        * in die nieuwe relaties kan je natuurlijk andere eerder gedefinieerde relaties weer gebruiken.
* **Voorbeelden**
* *Beperking:* 
    * Iedereen is man of vrouw (en nooit allebei tegelijk).
    * Iedereen heeft exact 2 ouders, een man en een vrouw.
* *Nieuwe relatie*
    * $Vader(x,y)$ precies dan en slechts dan alleen als x een ouder is van y en x een man is.
        * $\forall xy( Vader(x,y) \leftrightarrow R(x,y) \wedge M(x))$
    * $Opa(x,y)$ hiervoor gebruik je natuurlijk het $\exists z$    connectief.
        * Je kunt dit netjes definieren in termen van ouder en vader.

We begonnen met het maken van een overzicht van de familierelaties (Zie hierboven). Met behulp van dit overzicht konden we de relaties in logica definiÃ«ren.

 * 1. Familierelaties:
   * Vader: $\forall xy (Vader(x,y) \leftrightarrow (R(x,y) \wedge M(x)))$
   * Moeder: $\forall xy (Moeder(x,y) \leftrightarrow (R(x,y) \wedge V(x)))$

   * Broer: $ \forall xy (Broer(x,y) \leftrightarrow M(x) \wedge \exists z(R(x,z) \wedge R(y,z) \wedge V(z)))$
   * Zus: $ \forall xy (Zus(x,y) \leftrightarrow V(x) \wedge \exists z(R(x,z) \wedge R(y,z) \wedge V(z)))$

   * Opa: $\forall xy (Opa(x,y) \leftrightarrow (M(x) \wedge \exists z (R(x,z) \wedge R(z,y) )))$
   * Oma: $\forall xy (Oma(x,y) \leftrightarrow (V(x) \wedge \exists z (R(x,z) \wedge R(z,y) )))$

   * Oom (Broer van vader/moeder): $\forall xy (Oom(x,y) \leftrightarrow (M(x) \wedge \exists z (Broer(z,x) \vee Zus(z,x)) \wedge (Moeder(z,y) \vee Vader(z,y))))$
   * Oom (Aangetrouwde oom mits deze oom een kind heeft met de zus van vader/moeder): $\forall xy (Oom(x,y) \leftrightarrow (M(x) \wedge \exists zuv (Moeder(z,y) \vee Vader(z,y)) \wedge (Zus(u,z) \vee Broer(u,z)) \wedge (Moeder(u,v) \vee Vader(u,v)) \wedge (Vader(x,v))))$

   * Tante (Zus van vader/moeder): $\forall xy (Tante(x,y) \leftrightarrow (V(x) \wedge \exists z (Broer(z,x) \vee Zus(z,x)) \wedge (Moeder(z,y) \vee Vader(z,y))))$
   * Tante (Aangetrouwde tante mits deze tante een kind heeft met de broer van vader/moeder): $\forall xy (Tante(x,y) \leftrightarrow (M(x) \wedge \exists zuv (Moeder(z,y) \vee Vader(z,y)) \wedge (Zus(u,z) \vee Broer(u,z)) \wedge (Moeder(u,v) \vee Vader(u,v)) \wedge (Moeder(x,v))))$

   * Neef: $\forall xy (Neef(x,y) \leftrightarrow (M(x) \wedge (Oom(y,x) \vee Tante(y,x))))$
   * Nicht: $\forall xy (Nicht(x,y) \leftrightarrow (V(x) \wedge (Oom(y,x) \vee Tante(y,x))))$
   
   * OverGrootVader: $\forall xy (OverGrootVader(x,y) \leftrightarrow (M(x) \wedge \exists z (Opa(z,y) \vee Oma(z,y) ) \wedge (Vader(x,z))))$

   * OverGrootMoeder: $\forall xy (OverGrootVader(x,y) \leftrightarrow (V(x) \wedge \exists z (Opa(z,y) \vee Oma(z,y) ) \wedge Moeder(x,z)))$

   * OudOom: $\forall xy (OudOom(x,y) \leftrightarrow (M(x) \wedge \exists z ((Opa(z,y) \vee (Oma(z,y)) \wedge (Broer(x,z))))))$

   * Oudtante: $\forall xy (OudTante(x,y) \leftrightarrow (V(x) \wedge \exists z ((opa(z,y) \vee (Oma(z,y)) \wedge (Zus(x,z))))))$

   * AchterNeef: $\forall xy (AchterNeef(x,y) \leftrightarrow (M(x) \wedge \exists z ((OudOom(z,y) \vee OudTante(z,y)) \wedge (Vader(z,x) \vee Moeder(z,x)))))$
    
   * AchterNicht: $\forall xy (AchterNicht(x,y) \leftrightarrow (V(x) \wedge \exists z ((OudOom(z,y) \vee OudTante(z,y)) \wedge (Vader(z,x) \vee Moeder(z,x)))))$

   * HalfBroer: $\forall xy(HalfBroer(x,y) \leftrightarrow  (M(x) \wedge \exists zu (Vader(z,x) \wedge moeder(u,x)) \wedge ((\neg Vader(z,y) \wedge Moeder(u,y)) \vee (Vader(z,y) \wedge \neg Moeder(u,y)))))$

   * HalfZus: $\forall xy(Zus(x,y) \leftrightarrow  (V(x) \wedge \exists zu (Vader(z,x) \wedge moeder(u,x)) \wedge ((\neg Vader(z,y) \wedge Moeder(u,y)) \vee (Vader(z,y) \wedge \neg Moeder(u,y)))))$

```python
''' Comprehensies die de familierelaties teruggeven '''
def Man(x, network=G):
    return network.nodes[x]['M']
def Vrouw(x, network=G):
    return network.nodes[x]['V']
def R(x,y,network=G):
    return network.has_edge(x,y)

# Neemt een nx.Graph als input en geeft een DataFrame met alle familierelaties terug voor die Graph
def Relaties(S, relatie = ['Vader','Moeder','Broer','Zus','Opa','Oma','Neef','Nicht','Oom','Tante']):
  Vader= {(x,y) for x in S for y in S if R(x,y,S) and Man(x,S)}
  Moeder= {(x,y) for x in S for y in S if R(x,y,S) and Vrouw(x,S)}

  Zoon = {(x,y) for x in S for y in S if ((y,x) in Vader or (y,x) in Moeder) and Man(x,S)}
  Dochter = {(x,y) for x in S for y in S if ((y,x) in Vader or (y,x) in Moeder) and Vrouw(x,S) }

  Opa= {(x,y) for x in S for y in S if Man(x, S) and {z for z in S if R(x,z,S)} & {z for z in S if R(z,y,S)}} 
  Oma= {(x,y) for x in S for y in S if Vrouw(x, S) and {z for z in S if R(x,z,S)} & {z for z in S if R(z,y,S)}}

  # Halfbroer = {(x,y) for x in S if Man(x,S) for y in S if {z for z in S if R(z,x,S) and R(z,y,S) and x!=y}}
  # Halfzus = {(x,y) for x in S if Vrouw(x,S) for y in S if {z for z in S if R(z,x,S) and R(z,y,S) and x!=y}}

  Broer = {(x,y) for x in S if Man(x,S) for y in S if x!=y and {z for z in S if ((z,x) in Vader or (z,x) in Moeder) and ((z,y) in Vader or (z,y) in Moeder)}}
  Zus = {(x,y) for x in S if Vrouw(x,S) for y in S if x!=y and {z for z in S if ((z,x) in Vader or (z,x) in Moeder) and ((z,y) in Vader or (z,y) in Moeder)}}

  Oom = list({(x,y) for x in S if Man(x,S) for y in S if {z for z in S if ((z,x) in Broer or (z,x) in Zus) and ((z,y) in Moeder or (z,y) in Vader)}})
  Oom += {(x,y) for x in S if Man(x,S) for y in S if {z for z in S for u in S for v in S if ((z,y) in Moeder or (z,y) in Vader) and ((u,z) in Zus or (u,z) in Broer) and ((u,v) in Moeder or(u,v) in Vader) and ((x,v) in Vader)}}
    
  Tante = list({(x,y) for x in S if Vrouw(x,S) for y in S if {z for z in S if ((z,x) in Broer or (z,x) in Zus) and ((z,y) in Moeder or (z,y) in Vader)}})
  Tante += {(x,y) for x in S if Vrouw(x,S) for y in S if {z for z in S for u in S for v in S if ((z,y) in Moeder or (z,y) in Vader) and ((u,z) in Zus or (u,z) in Broer) and ((u,v) in Moeder or(u,v) in Vader) and ((x,v) in Vader)}}  
    
  Neef = {(x,y) for x in S if Man(x,S) for y in S if ((y,x) in Oom or (y,x) in Tante)}
  Nicht = {(x,y) for x in S if Vrouw(x,S) for y in S if ((y,x) in Oom or (y,x) in Tante)}

  
    
  d = {'Vader':list(Vader), 'Moeder':list(Moeder), 'Zoon':list(Zoon), 'Dochter':list(Dochter), 'Opa':list(Opa), 'Oma':list(Oma), 'Broer':list(Broer), 'Zus':list(Zus), 'Neef':list(Neef), 'Nicht':list(Nicht), 'Oom':list(Oom), 'Tante':list(Tante)}
  df = pd.DataFrame(dict([ (k,pd.Series(v, dtype='object')) for k,v in d.items() ]))
  df =df[relatie]

  return df

# De relaties van families kunnen opgeroepen worden met 'S', 'T', 'N' of 'MSS'.
Relaties(MSS)[:50]
```

Primitieven:

Man: $M$ 

Vrouw: $V$

Onzijdig: $N$

Transgender en intersex: Deze mensen identificeren zich als man of vrouw en vallen dus onder ook onder M en V. 

Sexuele voorkeur voor: $R(x,y)$

beperking: 

* Niemand heeft een seksuele voorkeur voor zichzelf: $\forall x (\neg R(x,x))$



* Heterosexueel:   $\forall xy (Hetero(x,y) \leftrightarrow ((M(x) \wedge V(y) \wedge R(x,y)) \vee (V(x) \wedge M(y) \wedge R(x,y)))) $

* Homoseksueel: $\forall xy (Homo(x,y) \leftrightarrow (M(xy) \wedge R(x,y))) $

* Lesbisch: $\forall xy (lesb(x,y) \leftrightarrow (V(xy) \wedge R(x,y))) $
    
* Biseksueel: $\forall xy (Bi(x,y) \leftrightarrow \exists z(((M(xz) \wedge V(y)) \wedge R(x,y) \wedge R(x,z)) \vee (V(xz) \wedge M(y)) \wedge R(x,y) \wedge R(x,z)))$

* Aseksueel: $\forall x(Aseks(x) \leftrightarrow \neg \exists y( R(x,y)))$

* Questioning: $\forall x,y(Quest(x,y) \leftrightarrow  (N(x) \wedge ((M(y) \vee V(y) \vee N(y)) \wedge R(x,y)))) $

* Panseksueel:  $\forall x,y(Pan(x,y) \leftrightarrow (M(x) \vee V(y) \wedge (Bi(x,y) \wedge \exists z( N(z) \wedge R(x,z)))))$

* Transgender: Een transgender identificeert zich als man of als vrouw en is dus in de context van relaties dus ook homo, lesbisch, hetero of biseksueel en kan daardoor niet gedefinieerd worden als aparte seksuele voorkeur.

* Intersex: Een persoon dat intersex is, is biologisch een man of vrouw maar vertoont geslachtskenmerken van het andere geslacht. Net als transgenders kiezen ze zelf als welk geslacht ze zich identificeren. Hierdoor zien we ze hier als man of vrouw en vallen ze niet onder een specifieke voorkeur.

```python
Q = nx.DiGraph()

Nodes = []

Man = {'Hetero_man', 'Homo_man', 'Bi_man', 'Aseks_man', 'Pan_man'}
Vrouw = {'Hetero_vrouw', 'Lesbi_vrouw', 'Bi_vrouw', 'Aseks_vrouw', 'Pan_vrouw'}
Ander = {'Questioning'}

for i in Man:
    Nodes.append(i)
for i in Vrouw:
    Nodes.append(i)
for i in Ander:
    Nodes.append(i)

Q.add_nodes_from(Nodes)

#assign color to genders
color_map = []
for node in Q:
    if node in Man:
        color_map.append('blue')
    elif node in Vrouw:
        color_map.append('pink')
    elif node in Ander:
        color_map.append('grey')
        
Q.add_edges_from([('Hetero_man', 'Hetero_vrouw'), ('Hetero_man', 'Lesbi_vrouw'), 
                  ('Hetero_man','Bi_vrouw'), ('Hetero_man', 'Aseks_vrouw'), ('Hetero_man', 'Pan_vrouw'),
                  ('Homo_man', 'Hetero_man'), ('Homo_man', 'Bi_man'),  
                  ('Homo_man', 'Aseks_man'), ('Homo_man', 'Pan_man'), ('Bi_man', 'Hetero_man'), 
                  ('Bi_man', 'Homo_man'), ('Bi_man', 'Aseks_man'), 
                  ('Bi_man', 'Hetero_vrouw'), ('Bi_man', 'Lesbi_vrouw'), 
                  ('Bi_man', 'Bi_vrouw'), ('Bi_man', 'Aseks_vrouw'), ('Bi_man', 'Pan_man'), ('Bi_man', 'Pan_vrouw'),
                  ('Hetero_vrouw', 'Hetero_man'), ('Hetero_vrouw', 'Homo_man'), 
                  ('Hetero_vrouw', 'Bi_man'), ('Hetero_vrouw', 'Aseks_man'), ('Hetero_vrouw', 'Pan_man'), 
                  ('Lesbi_vrouw', 'Hetero_vrouw'), ('Lesbi_vrouw', 'Bi_vrouw'), 
                  ('Lesbi_vrouw', 'Aseks_vrouw'), ('Lesbi_vrouw', 'Pan_vrouw'),
                  ('Bi_vrouw', 'Hetero_man'), ('Bi_vrouw', 'Homo_man'), ('Bi_vrouw', 'Bi_man'),
                  ('Bi_vrouw', 'Aseks_man'), ('Bi_vrouw', 'Pan_man'), ('Bi_vrouw', 'Hetero_vrouw'),
                  ('Bi_vrouw', 'Lesbi_vrouw'), ('Bi_vrouw', 'Aseks_vrouw'), ('Bi_vrouw', 'Pan_vrouw')
                 ])

#Questioning zegt niets over op wie diegene valt, dus diegene kan op iedereen vallen
for node in Nodes:
    if node not in Ander:
        Q.add_edge('Questioning', node)

for node in Nodes:   
    if node != 'Pan_vrouw':
        Q.add_edge('Pan_vrouw', node)
        
for node in Nodes:  
    if node != 'Pan_man':
        Q.add_edge('Pan_man', node)

plt.figure(4,figsize=(20,10))
nx.draw_circular(Q, with_labels=True, node_color=color_map, font_color='red', 
                 font_weight='bold', font_size = 20, arrowstyle='->', arrowsize=20, node_size=6000)
```
