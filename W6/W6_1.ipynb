# RDF en RDFS completion

* We maken een programma dat gegeven een RDF triple store plus een set RDFS statements, alle mogelijke niet triviale logische consequenties hiervan afleidt, en die in een RDF en RDFS database zet.
* Dit wordt ook wel *materialization* genoemd. Het idee is dat uit beide kennisbanken **exact dezelfde feiten volgen**, maar dat dat sneller kan vanuit de pure RDF versie (omdat je dan geen feiten hoeft af te leiden).
* De technische term voor het algorithme dat deze "completering" doet is de *chase*.
* Het [artikel van Enrico Franconi et al](https://link.springer.com/chapter/10.1007/978-3-642-41335-3_7) bevat een systeem.
* Net als in Franconi werken we alleen met de 4 interresantste RDFS statements:
    * **Hierarchie** 
        * $\forall xy (Rxy \rightarrow Sxy)$  (" to love him is to know him", als je van iemand houdt dan ken je diegene) 
        * $\forall x (Horse(x) \rightarrow Animal(x))$
    * **Domain en range restrictions**
        * $\forall xy (Rxy \rightarrow Dx)$ (domain) "Als x een staart heeft is het een dier"
        * $\forall xy (Rxy \rightarrow Dy)$ (range)  "Als je een  ouder bent van y, is y een mens."
        
## Jouw taak

    * voor elke feit `(o,p,s)`, facts+rules $\models$ (o,p,s) $\Leftrightarrow$ `complete(facts, rules)` $\models$ (o,p,s) 
        * Jouw functie is zowel volledig als geldig (*sound and complete*), dat is exact wat je wilt.
    * `complete(facts, rules)` is een verzameling feiten.

### Voorwaarden

#### Database en regels

We werken niet direct met RDF en RDFS omdat dat zo'n enorme syntax overhead heeft. We doen de dingen gewoon lekker in Python.

* Je RDF store is een verzameling triples, waarbij de 3 elementen strings of integers of floats zijn.
    * Je schrijft eigenschappen van individuen als `('henk', 'isa', 'Man')`
        * dus altijd met de vaste relatie `isa` in het midden.
    * en relaties tussen individuen als `('henk', Houdt_van', 'ingrid')`
    * en idem voor relaties tussen individuen en getallen `('henk', 'leeftijd', 53)`
    * Begin namen van individuen ("knopen") meteen kleine letter, en eigenschappen en relaties met een grote.
* Je RDFS store is een dict met sleutels `['subPropertyOf', 'subClassOf', 'domain', 'range']` en voor elke sleutel een verzameling met 2-tuples (paartjes in gewoon Nederlands) als waarde. 

#### Test voorbeelden

Een testset is een lijst triples van de vorm `(D:RDF_store, R:RDFS_store, Pos:RDF_store, Neg:RDF_store)` waarvoor geldt dat 
* voor elk triple t in Pos: $D,R\models t$
* voor elk triple t in Neg: $D,R\not\models t$
* Met $\models$ wordt dus bedoeld volgens de Franconi semantiek, en dus ook volgens de RDF semantiek. Dat betekent dus niet per se dat dit hetzelfde is als volgens jouw completion functie (want die zou fouten kunnen maken).

# Jouw completion functie

```python
studnr=11305320 # verander door het student id van jullie groepsleider
functienaam= 'complete_' + str(studnr)
print(functienaam)

def complete_11305320(facts, rules):  
    completion = facts
    
    while True:
        storeSize = len(completion)
        loopstore = {triple for triple in completion}
        
        for triple in loopstore:
            for subProp in rules['subPropertyOf']:
                if triple[1] == subProp[0]:
                    completion.add((triple[0], subProp[1], triple[2]))
                    
            for subClass in rules['subClassOf']:
                if triple[2] == subClass[0]:
                    completion.add((triple[0], triple[1], subClass[1]))
                    
            for dom in rules['domain']:
                if triple[1] == dom[0]:
                    completion.add((triple[0], 'isa', dom[1]))
                    
            for ran in rules['range']:
                if triple[1] == ran[0]:
                    completion.add((triple[2], 'isa', ran[1]))

        if storeSize == len(completion):
            break
            
    return completion

#test

functienaam = complete_11305320

complete_11305320(test1[0][0],test1[0][1])
```

# Voor de echte diehards die gaan voor een 10

Franconi beschrijft in zijn artikel dat je ook RDFS statements uit andere RDFS statements kunt afleiden. Hij geeft een leuk voorbeeld in het eerste plaatje.

Dat kan jij natuurlijk ook!

Maak een functie `complete(RDFS)` die precies alle RDFS statements afleidt die volgen uit de verzameling `RDFS`.

Maak er ook weer een mooie testset bij, en beargumenteer dat jouw systeem geldig en volledig is.

### Hoeveel RDFS statements zijn er eigenlijk die zouden kunnen volgen?

* Gegeven de relaties en eigenschappen in de verzameling `RDFS` is dit eindig.
* En er zijn maar 4 soorten RDFS statements. 
* Dus moet het wel eindig zijn, en kan je zelfs uitrekenen hoeveel.
* Een manier is om ze allemaal te proberen....
    * Voeg het antecedent toe als RDF-triple aan `RDFS` en kijk of het consequent dan volgt...
* Reken het eens uit voor $r$ relaties en $p$ eigenschappen.
    * Met $r$ relaties kan je $r\cdot(r-1)$ veel `subPropertyOf` relaties maken toch...
    * en dan nog ....
    * en ....
* Dit is niet eens zo gek, maar met een *chase* gaat het natuurlijk veel sneller omdat je dan resultaatgericht werkt in plaats van als een dolle alles gaat proberen.
    * Bijv; Stel je hebt 100 relaties, maar in je RDFS komt er maar 1 voor in antecdent positie in `subPropertyOf` statements. Hoeveel `subPropertyOf` kan je dan afleiden? Ja, dat is dus sneller dan al die 100*99 te gaan proberen toch? Want dat is best veel.

### RDFS chase

Naïeve versie:

Met $r$ relaties en $p$ eigenschappen kan je $r\cdot(r−1)$ veel `subPropertyOf` relaties maken

en dan nog $p\cdot(p−1)$ veel `subClassOf` relaties

en dan nog $2*r*p$ veel `domain` en `range` relaties samen.

---

Chase versie:
Stel je hebt 100 relaties, maar in je RDFS komt er maar 1 voor in antecdent positie in `subPropertyOf` statements. Hoeveel `subPropertyOf` je daarvan kan afleiden hangt af van de "regels" die je op de statements toepast, en  overlap met andere `range`, `domain` en `subClassOf` statements. Hoeveel je uiteindelijk uit die éne `subPropertyOf` hebt kunnen afleiden weet je pas aan het eind. Als je reflexiviteit toepast kan je iniedergeval ook gelijk afleiden dat het antecedent ook z'n eigen consequent in een nieuw statement is.

Het systeem is geldig en volledig omdat het letterlijk de regeltjes van het artikel in code bevat, en blijft loopen tot er niks meer afgeleid kan worden.

```python
from copy import deepcopy

def complete(RDFS):
    while True:
        storeSize = sum(len(RDFS[x]) for x in RDFS)
        loopstore = deepcopy(RDFS)
        
        for subPropA in loopstore['subPropertyOf']:
            # Regel 2b: subproperty transitiviteit: als A subproperty van B, dan A ook subproperty waar B subproperty van is
            for subPropB in loopstore['subPropertyOf']:
                if subPropA[1] == subPropB[0]:
                    RDFS['subPropertyOf'].add((subPropA[0], subPropB[1]))
                    
            # Regel 3b: Als A subprop van B is, neem domein waar B links staat over met A ipv B
            for dom in loopstore['domain']:
                if subPropA[1] == dom[0]:
                    RDFS['domain'].add((subPropA[0], dom[1]))
            
            # Regel 4b: Als A subprop van B is, neem range waar B links staat over met A ipv B
            for ran in loopstore['range']:
                if subPropA[1] == ran[0]:
                    RDFS['range'].add((subPropA[0], ran[1]))
                    
            # Regel 6b: Alles wat een subprop-relatie heeft is een subprop van zichzelf        
            RDFS['subPropertyOf'].add((subPropA[0], subPropA[0]))
            RDFS['subPropertyOf'].add((subPropA[1], subPropA[1]))

        for subClassA in loopstore['subClassOf']:
            # Regel 1b: subclass transitiviteit: als A subclass van B, dan A ook subclass waar B subclass van is
            for subClassB in loopstore['subClassOf']:
                if subClassA[1] == subClassB[0]:
                    RDFS['subClassOf'].add((subClassA[0], subClassB[1]))

            # Regel 3c: Als A domein van B is, neem domein van superclassen van B over        
            for dom in loopstore['domain']:
                if dom[1] == subClassA[0]:
                    RDFS['domain'].add((dom[0], subClassA[1]))
            
            # Regel 4c: Als A range van B is, neem range van superclassen van B over
            for ran in loopstore['range']:
                if ran[1] == subClassA[0]:
                    RDFS['range'].add((ran[0], subClassA[1]))
            
            # Regel 5a: Alles wat een subclass-relatie heeft is een subclass van zichzelf
            RDFS['subClassOf'].add((subClassA[0], subClassA[0]))
            RDFS['subClassOf'].add((subClassA[1], subClassA[1]))
            
        for ran in loopstore['range']:
            # Regel 5b: Ook alle classes in een range zijn subclass van zichzelf
            RDFS['subClassOf'].add((ran[1], ran[1]))
            
            # Regel 6d: Ook alle properties in een range zijn subprop van zichzelf
            RDFS['subPropertyOf'].add((ran[0], ran[0]))
            
        for dom in loopstore['domain']:
            # Regel 5b: Ook alle classes in een domein zijn subclass van zichzelf
            RDFS['subClassOf'].add((dom[1], dom[1]))
            
            # Regel 6d: Ook alle properties in een domein zijn subprop van zichzelf
            RDFS['subPropertyOf'].add((dom[0], dom[0]))
            
            # Regel 7: Extensional
            for subClassA in loopstore['subClassOf']:
                for subPropA in loopstore['subPropertyOf']:
                    if subPropA[0] == "isa" and subPropA[1] == dom[0] and subClassA[0] == subClassA[1]:
                        RDFS['subClassOf'].add((subClassA[0], dom[1]))
                
            
        if storeSize == sum(len(RDFS[x]) for x in RDFS):
            break
            
    return RDFS
```

```python
def testchecker_RDFS(tests, print_completion=False):
    for n, test in enumerate(tests):
        print(f"Test #{n}")
        completion = complete(test[0])
        for key in test[1]:
            for pos in test[1][key]:
                if pos not in completion[key]:
                    print(f"{pos} is not in completion, but should be!")
                else:
                    print(f"{pos} correctly in completion")
        for key in test[2]:
            for neg in test[2][key]:
                if neg in completion[key]:
                    print(f"{neg} is in completion, but shouldn't be!")
                else:
                    print(f"{neg} correctly not in completion")
        if print_completion:
            print("---")
            pprint(completion)
            print("---")
        print(f"#{n} Getest!\n---")
```
