# RDF en RDFS completion

* We maken een programma dat gegeven een RDF triple store plus een set RDFS statements, alle mogelijke niet triviale logische consequenties hiervan afleidt, en die in een RDF en RDFS database zet.
* Dit wordt ook wel *materialization* genoemd. Het idee is dat uit beide kennisbanken **exact dezelfde feiten volgen**, maar dat dat sneller kan vanuit de pure RDF versie (omdat je dan geen feiten hoeft af te leiden).
* De technische term voor het algorithme dat deze "completering" doet is de *chase*.
* Het [artikel van Enrico Franconi et al](https://link.springer.com/chapter/10.1007/978-3-642-41335-3_7) bevat een systeem.
* Net als in Franconi werken we alleen met de 4 interresantste RDFS statements:
    * **Hierarchie** 
        * $\forall xy (Rxy \rightarrow Sxy)$  (" to love him is to know him", als je van iemand houdt dan ken je diegene) 
        * $\forall x (Horse(x) \rightarrow Animal(x))$
    * **Domain en range restrictions**
        * $\forall xy (Rxy \rightarrow Dx)$ (domain) "Als x een staart heeft is het een dier"
        * $\forall xy (Rxy \rightarrow Dy)$ (range)  "Als je een  ouder bent van y, is y een mens."
        
## Jouw taak

1. Maak de functie `complete(facts, rules)` die voldoet aan de volgende eigenschap:
    * voor elke feit `(o,p,s)`, facts+rules $\models$ (o,p,s) $\Leftrightarrow$ `complete(facts, rules)` $\models$ (o,p,s) 
        * Jouw functie is zowel volledig als geldig (*sound and complete*), dat is exact wat je wilt.
    * `complete(facts, rules)` is een verzameling feiten.
2. Maak een flinke collectie van tests waarmee je test dat jouw functie aan die eisen voldoet. 
    * Dus dat het niet te veel en niet te weinig, maar exact de juiste feiten impliceert.
    * Je deelt die tests met alle anderen, en ook jouw functie, en we gaan dus elke functie tegen elke test draaien, en dus een matrix met scores maken.
    
### Je cijfer is gebaseerd op 

1. Hoe goed doorstaat jouw functie alle tests.
2. Hoe goed zijn jouw tests.
    * Strenge (doch rechtvaardige) tests zijn natuurlijk het best. Tests waar iedereen voor slaagt lijken leuk maar zijn meestal niet veel waard. 
    
    
### Voorwaarden

#### Database en regels

We werken niet direct met RDF en RDFS omdat dat zo'n enorme syntax overhead heeft. We doen de dingen gewoon lekker in Python.

* Je RDF store is een verzameling triples, waarbij de 3 elementen strings of integers of floats zijn.
    * Je schrijft eigenschappen van individuen als `('henk', 'isa', 'Man')`
        * dus altijd met de vaste relatie `isa` in het midden.
    * en relaties tussen individuen als `('henk', Houdt_van', 'ingrid')`
    * en idem voor relaties tussen individuen en getallen `('henk', 'leeftijd', 53)`
    * Begin namen van individuen ("knopen") meteen kleine letter, en eigenschappen en relaties met een grote.
* Je RDFS store is een dict met sleutels `['subPropertyOf', 'subClassOf', 'domain', 'range']` en voor elke sleutel een verzameling met 2-tuples (paartjes in gewoon Nederlands) als waarde. 

#### Test voorbeelden

Een testset is een lijst triples van de vorm `(D:RDF_store, R:RDFS_store, Pos:RDF_store, Neg:RDF_store)` waarvoor geldt dat 
* voor elk triple t in Pos: $D,R\models t$
* voor elk triple t in Neg: $D,R\not\models t$
* Met $\models$ wordt dus bedoeld volgens de Franconi semantiek, en dus ook volgens de RDF semantiek. Dat betekent dus niet per se dat dit hetzelfde is als volgens jouw completion functie (want die zou fouten kunnen maken).

# Jouw completion functie

```python
studnr=12855723 # verander door het student id van jullie groepsleider
functienaam= 'complete_'+str(studnr)


# functie die gegeven feiten en regels, itereert over de regels en ALLE feiten die hier logisch op volgen teruggeeft
def complete_12855723(facts, rules):
    while True:
        # maak een set voor nieuwe triples
        new_triples = set()
        
        # voeg nieuwe triples toe als gevolg van subPropertyOf
        for constraint in rules['subPropertyOf']:
            for triple in facts:
                # vb. als x houdt_van y, dan x kent y
                if triple[1] == constraint[0] and not (triple[0], constraint[1], triple[2]) in facts:
                    new_triples.add((triple[0], constraint[1], triple[2]))

        # voeg nieuwe triples toe als gevolg van subClassOf
        for constraint in rules['subClassOf']:
            for triple in facts:
                # vb. als x is_a mens, dan x is_a dier
                if triple[2] == constraint[0] and not (triple[0], triple[1], constraint[1]) in facts:
                    new_triples.add((triple[0], triple[1], constraint[1]))
        
        # voeg nieuwe triples toe als gevolg van domain
        for constraint in rules['domain']:
            for triple in facts:
                # vb. als x houdt_van y, dan x is_a mens
                if triple[1] == constraint[0] and not (triple[0], 'isa', constraint[1]) in facts:
                    new_triples.add((triple[0], 'isa', constraint[1]))
        
        # voeg nieuwe triples toe als gevolg van range
        for constraint in rules['range']:
            for triple in facts:
                # vb. als x vind_lekker y, dan y is_a eten
                if triple[1] == constraint[0] and not (triple[2], 'isa', constraint[1]) in facts:
                    new_triples.add((triple[2], 'isa', constraint[1]))
        
        # controleer of er nieuwe triples zijn gevonden, en stop als die er niet zijn
        if not new_triples:
            return facts
        # voeg deze triples samen met de bestaande
        facts = facts.union(new_triples)

        
# functie die gegeven een lijst met (rdf en rdfs) stores de logische gevolgen a.d.h.v. rdfs toevoegt aan de store 
def update_storelist(storelist):
    # maak een lijst voor geupdatete stores
    updated_stores = []
    
    for store in storelist:
        # nieuwe feiten verzamelen per store
        facts = complete_12855723(store[0], store[1])
        # geupdatete feiten in de lijst voegen
        updated_stores.append((facts, store[1], store[2], store[3]))
    
    # de lijst met geupdatete stores teruggeven
    return updated_stores


# functie die gegeven een lijst met tests teruggeeft of er per store aan de test voldaan wordt
def test_check(storelist):
    # update alle stores in de lijst
    storelist = update_storelist(storelist)
    # maak een dict voor de resultaten
    results = dict()
    
    # zoek voor overtreding per store
    for i, store in enumerate(storelist): 
        example_check = []    
        
        # check de positieve voorbeelden
        for triple in store[2]:
            if triple not in store[0]:
                example_check.append(False)
        
        # check de negatieve voorbeelden
        for triple in store[3]:
            if triple in store[0]:
                example_check.append(False)
        
        # voeg false toe als er een overtreding is gevonden
        if False in example_check:
            results["Test "+ str(i+1)] = False
        # en voeg true toe als er geen overtreding is gevonden
        else:
            results["Test "+ str(i+1)] = True
    
    # geef de dict terug met het resultaat per store
    return results


#test
test1, update_storelist(test1), test_check(test1)
# complete_12855723(test1[0][0],test1[0][1])
```

### Hoeveel RDFS statements zijn er eigenlijk die zouden kunnen volgen?

* Gegeven de relaties en eigenschappen in de verzameling `RDFS` is dit eindig.
* En er zijn maar 4 soorten RDFS statements. 
* Dus moet het wel eindig zijn, en kan je zelfs uitrekenen hoeveel.
* Een manier is om ze allemaal te proberen....
    * Voeg het antecedent toe als RDF-triple aan `RDFS` en kijk of het consequent dan volgt...
* Reken het eens uit voor $r$ relaties en $p$ eigenschappen.
    * Met $r$ relaties kan je $r\cdot(r-1)$ veel `subPropertyOf` relaties maken toch...
    * en dan nog ....
    * en ....
* Dit is niet eens zo gek, maar met een *chase* gaat het natuurlijk veel sneller omdat je dan resultaatgericht werkt in plaats van als een dolle alles gaat proberen.
    * Bijv; Stel je hebt 100 relaties, maar in je RDFS komt er maar 1 voor in antecdent positie in `subPropertyOf` statements. Hoeveel `subPropertyOf` kan je dan afleiden? Ja, dat is dus sneller dan al die 100*99 te gaan proberen toch? Want dat is best veel.

```python
# functie die alle RDFS statements uit een gegeven RDFS afleidt
def complete(RDFS):    
    # alle subproperties toevoegen
    while True:
        sub1 = {(a, y) for a,b in RDFS['subPropertyOf'] for x,y in RDFS['subPropertyOf'] if b == x}
        sub2 = {(a, a) for a,b in RDFS['subPropertyOf']}
        sub3 = {(b, b) for a,b in RDFS['subPropertyOf']}
        if not sub1.issubset(RDFS['subPropertyOf']):
            RDFS['subPropertyOf'].update(sub1)
        elif not sub2.issubset(RDFS['subPropertyOf']):
            RDFS['subPropertyOf'].update(sub2)
        elif not sub3.issubset(RDFS['subPropertyOf']):
            RDFS['subPropertyOf'].update(sub3)
        else:
            break

    # alle subclasses toevoegen
    while True:
        sub1 = {(a, y) for a,b in RDFS['subClassOf'] for x,y in RDFS['subClassOf'] if b == x}
        sub2 = {(a, a) for a,b in RDFS['subClassOf']}
        sub3 = {(b, b) for a,b in RDFS['subClassOf']}
        if not sub1.issubset(RDFS['subClassOf']):
            RDFS['subClassOf'].update(sub1)
        elif not sub2.issubset(RDFS['subClassOf']):
            RDFS['subClassOf'].update(sub2)
        elif not sub3.issubset(RDFS['subClassOf']):
            RDFS['subClassOf'].update(sub3)
        else:
            break
    
    # alle domeinen toevoegen
    while True:
        sub1 = {(y, b) for x,y in RDFS['subPropertyOf'] for a,b in RDFS['domain'] if a == x}
        sub2 = {(a, y) for x,y in RDFS['subClassOf'] for a,b in RDFS['domain'] if b == x}
        sub3 = {(a, a) for a,b in RDFS['domain']}
        sub4 = {(b, b) for a,b in RDFS['domain']}
        if not sub1.issubset(RDFS['domain']):
            RDFS['domain'].update(sub1)
        elif not sub2.issubset(RDFS['domain']):
            RDFS['domain'].update(sub2)
        elif not sub3.issubset(RDFS['subPropertyOf']):
            RDFS['subPropertyOf'].update(sub3)
        elif not sub4.issubset(RDFS['subClassOf']):
            RDFS['subClassOf'].update(sub4)
        else:
            break
    
    # alle ranges aan de hand van subproperties toevoegen
    while True:
        sub1 = {(y, b) for x,y in RDFS['subPropertyOf'] for a,b in RDFS['range'] if a == x}
        sub2 = {(a, y) for x,y in RDFS['subClassOf'] for a,b in RDFS['range'] if b == x}
        sub3 = {(a, a) for a,b in RDFS['range']}
        sub4 = {(b, b) for a,b in RDFS['range']}
        if not sub1.issubset(RDFS['range']):
            RDFS['range'].update(sub1)
        elif not sub2.issubset(RDFS['range']):
            RDFS['range'].update(sub2)
        elif not sub3.issubset(RDFS['subPropertyOf']):
            RDFS['subPropertyOf'].update(sub3)
        elif not sub4.issubset(RDFS['subClassOf']):
            RDFS['subClassOf'].update(sub4)
        else:
            break
    
    # nieuwe schema teruggeven
    return RDFS
```

```python
# testen   

# subproperty testen
t1 = ({'subPropertyOf': {('Geniet', 'Ervaart'), ('Eet','Vindt_lekker'), ('Vindt_lekker', 'Geniet')},
      'subClassOf': set(),
      'domain': set(),
      'range': set()},
     {('Eet', 'Ervaart')},
     {('Ervaart', 'Geniet')})

# subclass testen
t2 = ({'subPropertyOf': set(),
      'subClassOf': {('Mens','Dier'), ('Student','Mens'), ('Dier', 'Organisme')},
      'domain': set(),
      'range': set()},
     {('Mens', 'Organisme'), ('Student', 'Organisme')},
     {('Dier', 'Student')})

# domain testen aan de hand van subproperty
t3 = ({'subPropertyOf': {('Houdt_van', 'Kent')},
      'subClassOf': set(),
      'domain': {('Houdt_van','Mens'), ('Kent', 'Dier')},
      'range': set()},
     {('Kent', 'Mens')},
     {('Houdt_van', 'Dier')})

# domain testen aan de hand van subclass
t4 = ({'subPropertyOf': set(),
      'subClassOf': {('Mens', 'Dier'), ('Dier', 'Organisme')},
      'domain': {('Houdt_van','Dier')},
      'range': set()},
     {('Houdt_van', 'Organisme')},
     {('Houdt_van', 'Mens')})

# range testen aan de hand van subproperty
t5 = ({'subPropertyOf': {('Geniet', 'Ervaart'), ('Eet','Vindt_lekker'), ('Vindt_lekker', 'Geniet')},
      'subClassOf': set(),
      'domain': set(),
      'range': {('Vindt_lekker','Groente')}},
     {('Ervaart', 'Groente')},
     {('Eet', 'Groente')})

# range testen aan de hand van subclass
t6 = ({'subPropertyOf': set(),
      'subClassOf': {('Broccoli', 'Groente'), ('Groente', 'Eten')},
      'domain': set(),
      'range': {('Vindt_lekker','Groente')}},
     {('Vindt_lekker', 'Eten')},
     {('Vindt_lekker', 'Broccoli')})

# lege testset testen
t7 = ({'subPropertyOf': set(),
      'subClassOf': set(),
      'domain': set(),
      'range': set()},
     set(),
     set())

# subproperty reflexiviteit testen
t8 = ({'subPropertyOf': {('Houdt_van', 'Kent')},
      'subClassOf': set(),
      'domain': set(),
      'range': set()},
     {('Houdt_van', 'Houdt_van'), ('Kent', 'Kent')},
     {('Kent', 'Houdt_van')})

# subclass reflexiviteit testen
t9 = ({'subPropertyOf': set(),
      'subClassOf': {('Mens', 'Dier')},
      'domain': set(),
      'range': set()},
     {('Dier', 'Dier'), ('Mens', 'Mens')},
     {('Dier', 'Mens')})

# reflexiviteit in range en domein testen
t10 = ({'subPropertyOf': set(),
      'subClassOf': set(),
      'domain': {('Houdt_van', 'Mens')},
      'range': {('Vindt_lekker', 'Groente')}},
     {('Vindt_lekker', 'Vindt_lekker'), ('Mens', 'Mens')},
     {('Vindt_lekker', 'Mens')})

# RDFS zonder regels testen
t11 = ({'subPropertyOf': set(),
      'subClassOf': set(),
      'domain': set(),
      'range': set()},
     {('Mens', 'Mens'), ('Dier', 'Dier')},
     {('Mens', 'Dier')})

tests = [t1, t2, t3, t4, t5, t6, t7, t8, t9, t10, t11]

# functie testen
def test_complete(tests):
    for t in tests:
        c = complete(t[0])
        if list(t[0].values())[0:4] == [set(), set(), set(), set()]:
            c['subPropertyOf'] = t[1]
        values = [c['subPropertyOf'], c['subClassOf'], c['domain'], c['range']]
        values = set().union(*values)
        if t[1] == set() and t[2] == set():
            print('Er woorden geen testwaardes meegegeven')
        for answer in t[1]:
            if answer in values:
                print(answer, ': Het goede antwoord komt wel voor')
            else:
                print(answer, ': Helaas, het goede antwoord komt niet voor')
        for answer in t[2]:
            if answer in values:
                print(answer, ': Helaas, het foute antwoord komt wel voor')
            else:
                print(answer, ': Het foute antwoord komt niet voor')
        print(' ')

test_complete(tests)
```
