# RDF en RDFS completion

* We maken een programma dat gegeven een RDF triple store plus een set RDFS statements, alle mogelijke niet triviale logische consequenties hiervan afleidt, en die in een RDF en RDFS database zet.
* Dit wordt ook wel *materialization* genoemd. Het idee is dat uit beide kennisbanken **exact dezelfde feiten volgen**, maar dat dat sneller kan vanuit de pure RDF versie (omdat je dan geen feiten hoeft af te leiden).
* De technische term voor het algorithme dat deze "completering" doet is de *chase*.
* Het [artikel van Enrico Franconi et al](https://link.springer.com/chapter/10.1007/978-3-642-41335-3_7) bevat een systeem.
* Net als in Franconi werken we alleen met de 4 interresantste RDFS statements:
    * **Hierarchie** 
        * $\forall xy (Rxy \rightarrow Sxy)$  (" to love him is to know him", als je van iemand houdt dan ken je diegene) 
        * $\forall x (Horse(x) \rightarrow Animal(x))$
    * **Domain en range restrictions**
        * $\forall xy (Rxy \rightarrow Dx)$ (domain) "Als x een staart heeft is het een dier"
        * $\forall xy (Rxy \rightarrow Dy)$ (range)  "Als je een  ouder bent van y, is y een mens."
        
## Jouw taak

    * voor elke feit `(o,p,s)`, facts+rules $\models$ (o,p,s) $\Leftrightarrow$ `complete(facts, rules)` $\models$ (o,p,s) 
        * Jouw functie is zowel volledig als geldig (*sound and complete*), dat is exact wat je wilt.
    * `complete(facts, rules)` is een verzameling feiten.

### Voorwaarden

#### Database en regels

We werken niet direct met RDF en RDFS omdat dat zo'n enorme syntax overhead heeft. We doen de dingen gewoon lekker in Python.

* Je RDF store is een verzameling triples, waarbij de 3 elementen strings of integers of floats zijn.
    * Je schrijft eigenschappen van individuen als `('henk', 'isa', 'Man')`
        * dus altijd met de vaste relatie `isa` in het midden.
    * en relaties tussen individuen als `('henk', Houdt_van', 'ingrid')`
    * en idem voor relaties tussen individuen en getallen `('henk', 'leeftijd', 53)`
    * Begin namen van individuen ("knopen") meteen kleine letter, en eigenschappen en relaties met een grote.
* Je RDFS store is een dict met sleutels `['subPropertyOf', 'subClassOf', 'domain', 'range']` en voor elke sleutel een verzameling met 2-tuples (paartjes in gewoon Nederlands) als waarde. 

#### Test voorbeelden

Een testset is een lijst triples van de vorm `(D:RDF_store, R:RDFS_store, Pos:RDF_store, Neg:RDF_store)` waarvoor geldt dat 
* voor elk triple t in Pos: $D,R\models t$
* voor elk triple t in Neg: $D,R\not\models t$
* Met $\models$ wordt dus bedoeld volgens de Franconi semantiek, en dus ook volgens de RDF semantiek. Dat betekent dus niet per se dat dit hetzelfde is als volgens jouw completion functie (want die zou fouten kunnen maken).

# Jouw completion functie

```python
studnr=12716979 # verander door het student id van jullie groepsleider
functienaam= 'complete_'+str(studnr)
print(functienaam)

def complete_12716979(facts, rules):
    '''Voor elke feit (o,p,s), facts+rules \models⊨ (o,p,s) \Leftrightarrow ⇔ complete(facts, rules) \models⊨ (o,p,s)'''
    for fact in facts.copy():
        
        # split triple in object, predicate and subjcet
        o,p,s = fact[0], fact[1], fact[2]
        
        # loop through the different rules and restrictions
        for rule,restrictions in rules.items():
            for restriction in restrictions:
                
                # create dict with what triple has to be added with every rule
                rule_dict = {
                    "subPropertyOf" :[p, (o, restriction[1], s)],
                    "subClassOf" :  [s, (o, 'isa', restriction[1])],
                    "domain" : [p, (o, 'isa', restriction[1])],
                    "range" : [p, (s, 'isa', restriction[1])]
                }
                
                check = rule_dict[rule][0]
                triple = rule_dict[rule][1]
                
                # check if restriction of current triple is in the current rule
                if restriction[0] == check and triple not in facts:
                    facts.add(triple)
                    complete_12716979(facts, rules)
    return facts
    


#test
facts= test1[0][0]
rules= test1[0][1]
print(facts)
print(rules)
complete_12716979(facts,rules)
```

### Hoeveel RDFS statements zijn er eigenlijk die zouden kunnen volgen?

* Gegeven de relaties en eigenschappen in de verzameling `RDFS` is dit eindig.
* En er zijn maar 4 soorten RDFS statements. 
* Dus moet het wel eindig zijn, en kan je zelfs uitrekenen hoeveel.
* Een manier is om ze allemaal te proberen....
    * Voeg het antecedent toe als RDF-triple aan `RDFS` en kijk of het consequent dan volgt...
* Reken het eens uit voor $r$ relaties en $p$ eigenschappen.
    * Met $r$ relaties kan je $r\cdot(r-1)$ veel `subPropertyOf` relaties maken toch...
    * en dan nog ....
    * en ....
* Dit is niet eens zo gek, maar met een *chase* gaat het natuurlijk veel sneller omdat je dan resultaatgericht werkt in plaats van als een dolle alles gaat proberen.
    * Bijv; Stel je hebt 100 relaties, maar in je RDFS komt er maar 1 voor in antecdent positie in `subPropertyOf` statements. Hoeveel `subPropertyOf` kan je dan afleiden? Ja, dat is dus sneller dan al die 100*99 te gaan proberen toch? Want dat is best veel.

```python
def complete(RDFS):
    
    # Loop through the different rules and restrictions
    for rule, restrictions in RDFS.items():

        # Domain influences subClassOf, subPropertyOf influences range
        rule_dict = {
            "subPropertyOf" : restrictions.copy(),
            "subClassOf" : restrictions.copy(),
            "domain" : RDFS['subClassOf'],
            "range" : RDFS['subPropertyOf']
        }
        
        # Loop through the current restrictions and the correct dependent restrictions
        for (o, s) in restrictions.copy():
            for (o2, s2) in rule_dict[rule]:
                
                # Create the right checks for the right rules
                check_dict = {
                    "subPropertyOf" : [(s == o2), (o, s2)],
                    "subClassOf" : [(s == o2), (o, s2)],
                    "domain" : [(s == o2), (o, s2)],
                    "range" : [(o == s2), (o2, s)]
                }
                checks = check_dict[rule]
                
                # make sure restriction is not yet in RDFS list
                if checks[0] and checks[1] not in RDFS[rule]:
                    RDFS[rule].add(checks[1])
                    complete(RDFS)
    return RDFS

# Test function
def test_functie2(test):
    completion = complete(test[0])
    print('\n', completion, '\n')
    for rule, values in completion.items():
        pos_list = [p for p in test[1][rule] if p not in values]
        print(f'All positive examples {rule}: ', len(pos_list) == 0)
        neg_list = [p for p in test[2][rule] if p in values]
        print(f'No negative examples {rule}: ', len(neg_list) == 0)
        print('')


# Test lists in format rdfs, positives, negatives
test_rdfs1 = (
    {
        'subPropertyOf': {('Houdt_van','Is_vrienden_met'), ('Is_vrienden_met', 'Kent'), ('Eet','Vindt_lekker')},
        'subClassOf': {('Mens','Dier'),('Student','Mens')},
        'domain': {('Houdt_van','Mens')},
        'range': {('Vindt_lekker','Eten')}
    },
    {
        'subPropertyOf': {('Houdt_van', 'Kent')},
        'subClassOf': set(),
        'domain': {('Houdt_van', 'Mens')},
        'range': {('Eet', 'Eten')}
    },
    {
        'subPropertyOf': {('Is_vrienden_ment', 'Houdt_van')},
        'subClassOf': set(),
        'domain': ('Houdt_van', 'Student'),
        'range': {'Vindt_lekker', 'Dier'}
    }
)

test_rdfs2 = (
    {
    'subPropertyOf': {('Houdt_van', 'Kent'), ('Eet', 'Vindt_lekker')},
    'subClassOf': {('Mens', 'Omnivoor'), ('Omnivoor', 'Wezen')},
    'domain': {('Praat_met', 'Mens'), ('Praat_met', 'Wezen'), ('Vindt_lekker', 'Mens')},
    'range': {('Vindt_lekker', 'Eten')}
    },
    {
        'subPropertyOf': {('Houdt_van', 'Kent')},
        'subClassOf': {('Mens', 'Wezen')},
        'domain': {('Praat_met', 'Omnivoor')},
        'range': {('Eet', 'Eten')}
    },
    {
        'subPropertyOf': {('Is_vrienden_met', 'Houdt_van')},
        'subClassOf': set(),
        'domain': set(),
        'range': {'Vindt_lekker', 'Dier'}
    }
)

rdfs_tests = [test_rdfs1, test_rdfs2]

for i,test in enumerate(rdfs_tests):
    print(f"Test {i + 1}:")
    test_functie2(test)
    print("")
```

# Uitleg code
De gemaakte code voor het aanvullen van de RFDS statements is correct omdat de juiste output wordt weergegeven. De code vult de juiste "subPropertyOf" aan. Zo wordt de tuple ('houdt_van', 'kent') toegevoegd in test1 aan de hand van de tuples ('houdt_van', 'is_vrienden_met') & ('is_vrienden_met', 'kent'). Bij 'subClassOf' wordt in test 1 ook juist de tuple ('Student', 'Dier') aan de hand van de gegeven tuples ('Student', 'Mens') & ('Mens', 'Dier'). "subPropertyOf" en "subClassOf" hebben dus direct invloed op zichzelf, "domain" en "range" werken daarentegen op een iets andere manier. "subClassOf" heeft namelijk invloed op "domain", want zoals je ziet in test1, als je de subclass restricties ("Mens", "Dier") & ("Student", "Mens") en de domain restrictie ("Houdt_van", "Mens"), dan kan ("Houdt_van", "Dier") als domain restrictie toegevoegd worden, omdat Mens dus subclass is van Dier. Andersom kan ("Houdt_van", "Student") niet toegevoegd worden als domain restrictie, omdat deze niet direct afgeleid kan worden uit de restrictie ("Houdt_van", "Mens"). Als x van y houdt is x een mens, maar hoeft x dus niet per se een student te zijn, maar wel zeker een dier, omdat mens subclass is van mens. Hiernaast heeft "subPropertyOf" invloed op "range". Laten we weer kijken naar test1, deze test heeft de "subPropertyOf" restrictie ("Eet", "Vindt_lekker"), dus als je de triple (x, "Eet", pizza) hebt, dan kan je afleiden (x, "Vindt_lekker", pizza). Test1 heeft ook de "range" restrictie ("Vindt_lekker", "Eten"), dus als (x, "Vindt_lekker", pizza) dan (pizza, "isa", "Eten"). Uit deze twee restricties kan dan ook de "range" restrictie ("Eet", "Eten") afgeleid worden, dus zo kan je direct uit de triple (x, "Eet", pizza) de triple (pizza "isa" "Eten") afleiden.
