# Een probleem oplossen met WikiData, SparQL en Pandas
## Introductie

Met de WikiData webtool kun je door middel van SparQL Queries informatie opvragen over bijna alles wat gedocumenteerd is. In deze tutorial zullen we aan de hand van een voorbeeld uitleggen hoe je deze informatie kunt vinden, exporteren naar python en kan bewerken met pandas. 

## Wat is WikiData?

WikiData is een gezamenlijke database, die werkt op basis van RDF technologie. Deze database wordt onderandere gebruikt om wikipedia te ondersteunen met gegevens. Je kunt de gegevens uit de database halen door Queries op te stellen in SparQL. 
Stel nou dat we willen weten welke politici er allemaal uit Nederland komen. Dan maken we een query op deze manier:

### 1. Specificeren

Eerst moeten we specificeren welke informatie we nodig hebben. Dat doen we met "SELECT". We zetten hier ook "DISTINCT" achter. "SELECT DISTINCT" zorgt ervoor dat we geen dubbele informatie krijgen. 
In ons voorbeeld willen we de namen van alle Nederlandse Tweede Kamerleden. Dit ziet er dan zo uit: 

SELECT DISTINCT ?item

### 2. Zoeken

Nu we hebben gespecificeerd welke informatie we nodig hebben, kunnen we gaan zoeken. In SparQL gebeurt dat met "WHERE", gevolgd door accolades  {}. Binnen deze accolades zet je wanneer de informatie waar is. Deze zinnen bestaan uit 3 delen, in deze volgorde: item, property, value. In ons voorbeeld willen we iedereen die Tweede Kamerlid is geweest. In normale taal vragen we dit op de volgende manier:

Item die tweede kamerlid is.


In SparQL geef je de dingen die je wilt weten weer met een vraagteken er voor. In ons voorbeeld is dat ?item. De dingen die we wel weten, zijn Ambt (beroep) en Tweede Kamerlid. In WikiData wordt gewerkt met speciale codes voor elk object.
Voor elke property zet je wdt: en dan een bepaalde code die bij de property hoort. De code van Ambt is P39. 
Wanneer we een object met een bepaalde ambt vraag je dus "?item wdt:P39". Nu moeten we nog specificeren welk beroep we zoeken. Voor een object of value zet je wd: en dan de code die bij Tweede Kamerlid hoort. Dit is Q18887908. Ook moet er achter elke vraag een . zodat SparQL weet waar de vraag eindigt.

De volledige vraag die we aan WikiData gaan stellen is dus:

?item wdt:P39 wd:Q18887908 .

De gehele Query ziet er nu als volgt uit:




```python
"""
SELECT DISTINCT ?item 
WHERE { 
?item wdt:P39 wd:Q18887908 .
}
"""
```

Klik op deze link om de query te runnen in de WikiData omgeving:

Je ziet dat de Query allemaal codes teruggeeft. Met deze informatie kunnen we nog weinig. Hier heeft WikiData wat op bedacht. Als je deze code toevoegt aan je query, zoekt WikiData de naam die bij de code hoort:

{

SERVICE wikibase:label { bd:serviceParam wikibase:language "nl" }

}

Wanneer je dit doet, moet je wel iets extra's toevoegen aan de "SELECT" vraag. Dit is de naam van het onderdeel wat je zoekt, met Label erachter. In ons voorbeeld zal de Query er dan zo uit zien:


```python
"""
SELECT DISTINCT ?item ?itemLabel
WHERE { 
?item wdt:P39 wd:Q18887908 .
      
{
SERVICE wikibase:label { bd:serviceParam wikibase:language "nl" }
}
}
"""
```

Wanneer je de Query goed hebt ontwikkeld, kunnen we beginnen aan het importeren van de gegevens naar Python. Dit gaat vrij simpel. De eerste stap is het importeren van de requests library met de onderstaande code

Zoals je ziet is C het juiste antwoord. Hier willen we iets aan doen. Gelukkig heeft SparQL hier een simpele optie voor bedacht, namelijk "OPTIONAL". Met deze functie kan er gezocht worden naar extra waardes, maar alleen als ze er zijn. Je zet de vraag die optioneel is binnen de accolades achter "OPTIONAL". Omdat volgers wordt teruggegeven als een getal hoeven we ?volgersLabel niet te gebruiken. Ook moeten we volgers toevoegen in het omzetten naar een dataset. In het voorbeeld hieronder kun je zien hoe we dat hebben gedaan. 


```python
query2 = """
SELECT DISTINCT ?item ?itemLabel ?school ?schoolLabel ?volgers 
WHERE { 
?item wdt:P39 wd:Q18887908 . 
?item wdt:P69 ?school .

OPTIONAL{
?item wdt:P8687 ?volgers .
}
      
{
SERVICE wikibase:label { bd:serviceParam wikibase:language "nl" }
}
}
"""
r = requests.get(url, params = {'format': 'json', 'query': query2})
data2 = r.json()

politicus = []
for item in data2['results']['bindings']:
    voegtoe = OrderedDict({
        'Naam':item['itemLabel']['value'],
        'Universiteit': item['schoolLabel']['value']
    })
    if 'volgers' in item :
        voegtoe['volgers'] = item['volgers']['value']

    politicus.append(voegtoe)
PolitiekDatabase = pd.DataFrame(politicus)
PolitiekDatabase.set_index('Naam', inplace=True)
PolitiekDatabase.head(10)


```

Om te beginnen gaan we weer door de wikidata met SPARQL om een nieuwe dataset te krijgen die de andere beroepen van de politici uit de Tweede Kamer weergeeft, en hoe vaak deze voorkomen. Hierbij wordt de COUNT() functie gebruikt, die telt hoe vaak een bepaalde variabele voorkomt in een gegroepeerde dataset. Zo wordt uiteindelijk de variabele en de frequentie meegegeven.

```python
url2 = 'https://query.wikidata.org/sparql'
query2 = """
SELECT ?occupationLabel (COUNT(?occupation) as ?Ocount)
WHERE { 

?item wdt:P39 wd:Q18887908 .
?item wdt:P106 ?occupation .
  
filter(?occupation != wd:Q82955)  

{
SERVICE wikibase:label { bd:serviceParam wikibase:language "nl" }
}
}
GROUP BY ?occupationLabel
"""
r2 = requests.get(url2, params = {'format': 'json', 'query': query2})
data2 = r2.json()

data2
```

Nu zetten we de data weer om tot een pandas dataframe met de volgende code:

```python
countries = []
for item in data2['results']['bindings']:
    countries.append(OrderedDict({
        'occupation': item['occupationLabel']['value'],
        'count':item['Ocount']['value']}))
df2 = pd.DataFrame(countries)
df2.set_index('occupation', inplace=True)
df2
```

De df2 database bestaat nu uit alle andere beroepen van de Nederlandse politici en de frequentie van hoe vaak deze beroepen voorkomen. Er is echter een groot aantal rijen waarbij maar een heel klein aantal van de politici heeft aangegeven dat beroep te hebben. Als hiervan een taartdiagram wordt gemaakt komen er een heleboel miniscule stukjes in die de figuur moeilijk leesbaar maakt. Deze kleine waardes kunnen samengevoegd worden in een categorie die 'anders' wordt genoemd, om al deze kleine stukjes samen te voegen tot de beroepen die weinig populariteit hebben onder de kamerleden.

```python
# Frequenties naar integers omzetten
df3 = pd.to_numeric(df2['count'], downcast='integer')

# Kleine en grote waardes onderscheiden en alle kleine waardes bij elkaar optellen en tot een nieuwe series vormen
small_values = df3.loc[df3.values < 15]
others = pd.Series(data={'anders': sum(small_values)})
big_values = df3.loc[df3.values >= 15]

# Waardes samenvoegen in nieuwe set
pievalues = pd.concat([big_values, others], axis=0)

pievalues
```

### 6.1 Een taartdiagram
Nu is het tijd om een figuur te maken van de data. Bij dit frequentie dataframe past een taartdiagram het beste om weer te geven wat de verdeling van de waardes is. Om te laten zien wat het verschil is in het bewerken van je data hebben wij met behulp van plotly graph objects (een uitgebreide versie van plotly express) een interactief figuur gemaakt.
```python
import plotly.graph_objs as go

# Taartdiagrammen maken van de data
fig1 = go.Pie(labels=df3.index, values=df3.values)
fig2 = go.Pie(labels=pievalues.index, values=pievalues.values, textposition='inside', textinfo='label')

data = [fig1, fig2]
group = go.Figure(data=data[0:2])

# Knoppen menu van de twee verschillende diagrammen aanmaken
group.update_layout(updatemenus=[dict(
    type = "buttons", direction = "left",
    buttons=list([dict(args=[{'visible':[False,True]},
                             {'title':'Heel veel waardes',
                             'annotations':[]}],
                       label="Bewerkt", method="update"),
                  dict(args=[{'visible':[True,False]},
                             {'title':'Samengevatte waardes',
                              'annotations':[]}], 
                       label="Onbewerkt", method="update"),]),active = 0, showactive=False,)])

group.show()
```

We kunnen ook een grote database gebruiken om figuren mee te maken, in plaats van een dataframe met maar twee kolommen. Zo kunnen er ook meer dan twee variabelen met elkaar vergeleken worden!

```python
#filter(?school = wd:Q214341 || ?school = wd:Q1065414)
url = 'https://query.wikidata.org/sparql'
query = """
SELECT DISTINCT ?item ?itemLabel ?schoolLabel ?partyLabel ?genderLabel
WHERE { 

?item wdt:P39 wd:Q18887908 .
?item wdt:P69 ?school .
?school wdt:P31 wd:Q3918 .
?item wdt:P21 ?gender .
?item wdt:P102 ?party .


{
SERVICE wikibase:label { bd:serviceParam wikibase:language "nl" }
}
}
"""
r = requests.get(url, params = {'format': 'json', 'query': query})
data = r.json()

countries = []
for item in data['results']['bindings']:
    countries.append(OrderedDict({
        'Naam':item['itemLabel']['value'],
        'uni': item['schoolLabel']['value'],
        'party': item['partyLabel']['value'],
        'geslacht': item['genderLabel']['value'],}))
df = pd.DataFrame(countries)
df.set_index('Naam', inplace=True)
df.head()
```

Aan de hand van deze mooie dataset kunnen wij nu ontzettend veel nieuwe diagrammen maken die ons telkens weer andere informatie weergeven. Een voorbeeld is een een diagram met de verdeling van de universiteiten waar alle politici uit de Tweede Kamer heen zijn gegaan:

In het volgende deel van de tutorial gaan we kijken of er politieke verschillen bestaan tussen universiteiten. Daarvoor wordt gebruikt gemaakt van iets lastigere functies zoals groupby, maar daarover later meer. Om te beginnen moeten we weer data ophalen bij Wikidata.

```python
url = 'https://query.wikidata.org/sparql'
query = """
SELECT DISTINCT 
?item ?itemLabel 
?school ?schoolLabel 
?partij ?partijLabel
?ideologie ?ideologieLabel
WHERE { 

?item wdt:P39 wd:Q18887908 .
?item wdt:P69 ?school .
?item wdt:P102 ?partij .
?partij wdt:P1142 ?ideologie .
{
SERVICE wikibase:label { bd:serviceParam wikibase:language "en" }
}
}
"""
r = requests.get(url, params = {'format': 'json', 'query': query})
data = r.json()
data
```

```python
countries = []
for item in data['results']['bindings']:
    countries.append(OrderedDict({
        'Naam':item['itemLabel']['value'],
        'Universiteit': item['schoolLabel']['value'],
        'Partij':item['partijLabel']['value'],
        'Ideologie':item['ideologieLabel']['value']
    
    }))
df = pd.DataFrame(countries)
df.set_index('Naam', inplace=True)
display(df.head(5))
```

#### Verwijderen dubbele personen
De vraag die we met deze data o.a. willen beantwoorden is hoeveel personen er van elke partij in deze data zijn opgenomen. Als een persoon op 5 verschillende universiteiten heeft gezeten, en maar bij een partij heeft gezeten, dan zou de data met dubbele waardes 5x deze partij meetellen. Dit willen we niet, want dat geeft een vertekend beeld, dus hebben we de dubbele waarders verwijderd. Bij het verwijderen van de data moet je keuzes maken, want er zijn misschien ook personen die wel bij 2 partijen actief zijn geweest. De functie die wij hebben gekozen houdt hier geen rekening mee, maar dit is een bewuste keuze, want er zijn waarschijnlijk niet heel veel gevallen van mensen die bij 2 partijen actief zijn geweest. 

We tonen beide plots hieronder.
Als je naar de y-as kijkt van beide plots dan zie je dat de aantallen van de personen enorm verschillen. Het heeft in deze situatie ook gevolgen voor de resultaten. Zonder dubbele personen is het CDA (Christian Democratic Appeal) de 3e partij. Nemen we echter personen die vaker voorkomen elke keer mee, dan is het CDA veruit de grootste partij.

```python
df_no_dub = df[~df.index.duplicated(keep='first')]
```

```python
threshold1 = 15

no_dub = df_no_dub.groupby(["Partij"])["Partij"].count().reset_index(name="count")
no_dub = no_dub[no_dub['count'] >= threshold1]

dub = df.groupby(["Partij"])["Partij"].count().reset_index(name="count")
dub = dub[dub['count'] >= threshold1]


fig = px.histogram(no_dub, x="Partij", y="count", title='Voorkomen partij zonder dubbele personen')
fig_dub = px.histogram(dub, x="Partij", y="count", title='Voorkomen partij met dubbele personen')
fig.show(),fig_dub.show()
display(no_dub.head(30))
```

#### We willen nu gaan kijken of bepaalde universiteiten bepaalde politieke stromingen vertegenwoordigen
Ook hier moeten we weer goed nadenken over wat de met de dubbele waardes gaan doen. Een persoon die op meerdere personen heeft gezeten, komt meerdere keren voor en de ideologie waar deze persoon voor staat telt zo mee voor de verschillende universiteiten. IN deze situatie hebben wij besloten om de dubbele waardes niet te verwijderen, omdat wij vinden dat de ideologie van een persoon op meerdere universiteiten mag meetellen.

We maken gebruik van relatieve percentages, omdat die een evenwichtiger beeld geeft als kleine en grote universiteiten met elkaar worden vergleken.


Een andere keuze die gemaakt moet worden is de ondergrens voor het aantal keer dat een stroming moet voorkomen op een universiteit. Hiervoor maken we gebruik van de slider hieronder: 
als de slider op 5 staat, dan moet een politieke ideologie minimaal 5x voorkomen op een universiteit. Deze keuze heeft grote invloed op de uiteindelijke grafiek. Probeer maar! (schuif eerste de slider, en run dan de cel eronder)

```python
import ipywidgets as widgets
threshold = widgets.IntSlider(max=50)
display(threshold)
```

```python
x = df.groupby(["Universiteit", "Ideologie"])["Ideologie"].count().reset_index(name="count")

x = x[x['count'] >= (threshold.value)]

y = x.groupby('Universiteit').sum()
y = y.to_dict()

x['totaal_uni'] = x['Universiteit'].map(y['count'])
x['Voorkomen_in_%'] = (x['count'] / x['totaal_uni'])*100

fig = px.bar(x, x="Universiteit", y="Voorkomen_in_%", color="Ideologie", title="Ideologie per universiteit")
fig.show()
display(x.head(5))
```
