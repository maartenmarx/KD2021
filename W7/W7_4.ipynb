# Liedjes per jaar

In muziekcredits wordt tegenwoordig meestal de distinctie gemaakt tussen schrijvers en producenten van het nummer, waar vroeger meestal de set componist en tekstschrijver (lyricist) werd gebruikt. Met een SPARQL-query kan je bijvoorbeeld kijken naar alle liedjes geschreven of geproduceerd door een Nederlander. Je checkt dan eerst de nationaliteit met Wikidata-attribuut `P27`, kijkt of iemand een een schijver van een werk is met `P86` (componist) of `P676` (tekstschrijver), deze worden in Wikidata ook wel eens door elkaar gebruikt. Met attribuut `P577` kunnen we ook nog achter de releasedatum van het betreffende nummer kijken. Of slechts het releasejaar, want niet altijd is de volledige datum ingevuld of bekend. Ook belangrijk dat we alleen kijken naar liedjes, dus we moeten kijken of het nummer wel echt een _song_ `Q7366` of een single `Q134556` is. Films kunnen namelijk ook gecomponeerd worden en soms staat een producer gekoppeld aan een heel album.

```python
query_wy = """SELECT DISTINCT ?writerLabel ?songLabel ?jaar
WHERE 
{
  ?writer wdt:P27 wd:Q29999. #writer is een Nederlander
  ?song (wdt:P86|wdt:P676) ?writer. #liedje als path
  ?song wdt:P577 ?jaar. #releasedatum
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""

query_py = """SELECT DISTINCT ?producerLabel ?songLabel ?jaar
WHERE 
{
  ?producer wdt:P27 wd:Q29999. #producer is een Nederlander
  ?song wdt:P162 ?producer. #liedje als path
  ?song wdt:P577 ?jaar. #releasedatum
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""

df_wy = sparql_dataframe.get(endpoint_url, query_wy, post=True)
df_py = sparql_dataframe.get(endpoint_url, query_py, post=True)

#Pakken het jaar uit het datetimeobject, veel liedjes hebben alleen het jaar precies ingevuld
df_wy["jaar"] = pd.DatetimeIndex(df_wy['jaar']).year
df_py["jaar"] = pd.DatetimeIndex(df_py['jaar']).year

display(df_wy, df_py)
```

## Aantallen visualiseren

De resultaten van de queries hebben we door middel van `sparql_dataframe` hierboven omgezet in `pandas`-dataframes, dan is er gelijk een stuk meer mogelijk! Zo hebben we alle datums al gelijk omgezet naar alleen het jaar. Als we een plotje willen maken van hoeveel liedjes elke Nederlandse schrijver elk jaar heeft geschreven moeten we nog even per jaar en per schrijver gaan optellen. Een liedje kan natuurlijk een co-geschreven werk zijn, dus één samengeschreven lied telt voor iedere schrijver als een lied. Met pandas kunnen we dat optellen doen, en ervoor zorgen dat als je over de `plotly` visualisatie heen gaat kunt zien wie dat jaar hoeveel liedjes geschreven/geproduceerd heeft.

```python
def hoverinfo(agg):
    df = pd.DataFrame(agg, columns = [agg.name])
    df['rank'] = df.groupby(agg.name).cumcount() + 1
    df = df.sort_values('rank').drop_duplicates([agg.name],keep='last').sort_values(by=['rank',agg.name],ascending=[False,True])
    return '<br>'.join([f"{row[agg.name]} ({row['rank']})" for index, row in df.iterrows()])

df_wy_grouped = df_wy.groupby(df_wy['jaar']).count()['writerLabel'].rename('Count').to_frame()
df_wy_grouped['schrijvers'] = df_wy.groupby(df_wy['jaar'])['writerLabel'].agg(hoverinfo)

df_py_grouped = df_py.groupby(df_py['jaar']).count()['producerLabel'].rename('Count').to_frame()
df_py_grouped['producenten'] = df_py.groupby(df_py['jaar'])['producerLabel'].agg(hoverinfo)

fig_wy = px.bar(
    df_wy_grouped, y='Count', title='<b>Aantal liedjes per Nederlandse songwriter in Wikidata, per jaar</b>',
    labels= {"Count": "Aantal", "jaar": "Jaar", "schrijvers": "Schrijvers"},
    hover_data=['schrijvers'])

fig_py = px.bar(
    df_py_grouped, y='Count', title='<b>Aantal liedjes per Nederlandse producent in Wikidata, per jaar</b>',
    labels= {"Count": "Aantal", "jaar": "Jaar", "producenten": "Producenten"},
    hover_data=['producenten'])

@interact(control=widgets.RadioButtons(
    options=["Schrijvers", "Producenten"],description='Groep:'))

def plot_fig(control):
    {"Schrijvers": fig_wy, "Producenten": fig_py}[control].show()
```

# Waar komen Nederlandse muzikanten vandaan?

Een andere interessante vraag waar je door middel van data uit Wikidata gedeeltelijk inzicht in kan krijgen is waar een hele groep mensen geboren is. Onderstaand doen we dat voor Nederlandse muziekschrijvers, producenten en artiesten (hier zit natuurlijk ook wat overlap in). Voor een persoon in Wikidata kan de property geboorteplaats `P19` ingevuld zijn, en met de coördinaten-property `P625` kunnen we daar dan weer de locatie van ophalen. 

Voor het gemak in de visualisatie hebben we de regel dat de geboorteplaats in Nederland moet liggen toegevoegd, dus Nederlandse muzikanten geboren in het buitenland doen deze keer niet mee. Sorry Dotan.

```python
dflist = []

for prop, name in zip(["(wdt:P86|wdt:P676)", "wdt:P162", "wdt:P175"], ["Schrijvers", "Producenten", "Artiesten"]):
    query = """SELECT DISTINCT ?personLabel ?gebplaatsLabel ?coord
    WHERE 
    {
      ?person wdt:P27 wd:Q29999. #persoon is een Nederlander
      ?person wdt:P19 ?gebplaats.
      ?gebplaats wdt:P17 wd:Q55. # Voor gemak visualisatie: ook alleen in NL geboren NL'ers
      ?gebplaats wdt:P625 ?coord.
      ?song """ + prop + """ ?person. #pad naar liedje vinden
      {
      ?song wdt:P31 wd:Q7366. #alleen songs
      }
      UNION {
      ?song wdt:P31 wd:Q134556 #en singles
      }
      SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
    }"""
    
    df = sparql_dataframe.get(endpoint_url, query, post=True)
    print(name+":")
    display(df.head())
    print(len(df), "rows")
    dflist.append(df)
```

## Markeren op de kaart

Om er daadwerkelijk pap van te maken is het logisch om per groep een kaartje te maken, en dat is exact wat we gaan doen. Eerst moeten we nog wat voorverwerking doen, hier in de vorm van optellen per stad hoeveel mensen daar geboren zijn, en de coördinatenkolom splitsen in lengtegraad en breedtegraad.

```python
@interact(plot=widgets.RadioButtons(
    options=["Schrijvers", "Producenten", "Artiesten"],description='Groep:'))

def plot_map(plot):
    df = dflist[{"Schrijvers": 0, "Producenten": 1, "Artiesten": 2}[plot]]
    
    # Extra verwerking voor plot
    df_grouped = df.groupby(['gebplaatsLabel','coord']).count()["personLabel"].rename('Count').to_frame().reset_index()
    df_grouped["coord"] = df_grouped["coord"].apply(lambda x: tuple(x[6:-1].split()))
    df_grouped[['lon','lat']] = pd.DataFrame(df_grouped.coord.to_list(), index= df_grouped.index)
    df_grouped['text'] = df_grouped['gebplaatsLabel'] + '<br>Aantal: ' + (df_grouped['Count']).astype(str)

    # Schaal voor datapunten
    scale = 2 * max(df_grouped["Count"]) / (9 ** 2)

    fig = go.Figure()
    fig.add_trace(go.Scattergeo(
        lon = df_grouped['lon'],
        lat = df_grouped['lat'],
        text = df_grouped['text'],
        marker = dict(
            size = (df_grouped["Count"]/scale)+1,
            color = "orange",
            line_color='rgb(40,40,40)',
            line_width=0.5,
        ),
        name = "Plaats"))

    fig.update_layout(
            title_text = f"Geboorteplaatsen van Nederlandse muziek{plot.lower()} in Wikidata" ,
            showlegend = True,
            geo = dict(
                resolution = 50,
                showcountries=True,
                projection = dict(type="mercator"),
                projection_scale=70, #this is kind of like zoom
                center=dict(lat=52.15, lon=5.38), # this will center on the point
            )
        )

    fig.show()
```

# Wie schrijft voor wie

Nog een vraag die ons erg interessant leek is: Wie schrijft voor wie?
In de query hebben we besloten om tekstschrijvers, componisten en producers te betrekken.

In de query hebben we twee restricties gemaakt. Er is voor gekozen om mensen die voor zichzelf schrijven eruit te halen. Het is voor ons niet heel interessant dat Duncan Laurence bijvoorbeeld meegeschreven heeft aan z'n eigen songfestivalinzending. Daarnaast zijn ook schrijvers die schrijver zijn voor de groep waar ze in zitten verwijdert. Dit is ook gedaan omdat het ons niet zo interessant leek. Helaas lukt Wikidata het niet altijd om deze eruit te filteren, omdat deze data niet altijd ingevoerd is.

Een belangrijk iets om te betrekken in de query is om alleen liedjes te selecteren omdat men ook voor een film of serie kan produceren en componeren.

```python
query = """SELECT DISTINCT ?itemLabel ?artistLabel ?songLabel
WHERE 
{
  ?item wdt:P27 wd:Q29999.
  ?song (wdt:P86|wdt:P676|wdt:P162) ?item.
  ?song wdt:P175 ?artist.
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  FILTER NOT EXISTS { # niet liedjes waar de artiest ook de schrijver is (Dus niet martin garrix als schrijver voor dua lipa als ze allebei uitvoerend artiest zijn)
     ?song wdt:P175 ?item.
     ?song (wdt:P86|wdt:P676|wdt:P162) ?item.
  }
  FILTER NOT EXISTS { # schrijvers die onderdeel zijn van de groep waarvoor ze schrijven eruit
    ?item wdt:P463 ?artist.
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}
"""

df = sparql_dataframe.get(endpoint_url, query, post=True)
df
```

```python
G = nx.DiGraph()
G.add_edges_from({(maker, uitvoerder) for maker, uitvoerder in zip(df['itemLabel'], df['artistLabel'])})
pos = nx.nx_agraph.graphviz_layout(G, prog="neato")

nx.draw(G, pos=pos, node_size=15)
```

## Nu nog interactief

Met dank aan code op github konden we pijltjes toevoegen in de plotly visualisatie, zodat het nogsteeds weergeven werd als gerichte graaf.

Het is een behoorlijke hoeveelheid code, maar de Networkx edges moeten worden overgezet worden naar Plotly lijnen. Dit houdt in dat de twee uiteinden van de edges worden omgezet naar coördinaten voor de Plotly grafiek.

```python
def addEdge(start, end, edge_x, edge_y, lengthFrac=1, arrowPos = None, arrowLength=0.025, arrowAngle = 30, dotSize=20):
    # Get start and end cartesian coordinates
    x0, y0 = start
    x1, y1 = end
    
    # Avoid zero division: Robert fix
    if (x0==x1):
        x1 += 0.01
    if (y0==y1):
        y1 += 0.01
    
    # Incorporate the fraction of this segment covered by a dot into total reduction
    length = math.sqrt( (x1-x0)**2 + (y1-y0)**2 )
    dotSizeConversion = .0565/20 # length units per dot size
    convertedDotDiameter = dotSize * dotSizeConversion
    lengthFracReduction = convertedDotDiameter / length
    lengthFrac = lengthFrac - lengthFracReduction
    
    # If the line segment should not cover the entire distance, get actual start and end coords
    skipX = (x1-x0)*(1-lengthFrac)
    skipY = (y1-y0)*(1-lengthFrac)
    x0 = x0 + skipX/2
    x1 = x1 - skipX/2
    y0 = y0 + skipY/2
    y1 = y1 - skipY/2
    
    # Append line corresponding to the edge
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None) # Prevents a line being drawn from end of this edge to start of next edge
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)
    
    # Draw arrow
    if not arrowPos == None:
        
        # Find the point of the arrow; assume is at end unless told middle
        pointx = x1
        pointy = y1
        eta = math.degrees(math.atan((x1-x0)/(y1-y0)))
        
        if arrowPos == 'middle' or arrowPos == 'mid':
            pointx = x0 + (x1-x0)/2
            pointy = y0 + (y1-y0)/2
            
        # Find the directions the arrows are pointing
        signx = (x1-x0)/abs(x1-x0)
        signy = (y1-y0)/abs(y1-y0)
        
        # Append first arrowhead
        dx = arrowLength * math.sin(math.radians(eta + arrowAngle))
        dy = arrowLength * math.cos(math.radians(eta + arrowAngle))    
        edge_x.append(pointx)
        edge_x.append(pointx - signx**2 * signy * dx)
        edge_x.append(None)
        edge_y.append(pointy)
        edge_y.append(pointy - signx**2 * signy * dy)
        edge_y.append(None)
        
        # And second arrowhead
        dx = arrowLength * math.sin(math.radians(eta - arrowAngle))
        dy = arrowLength * math.cos(math.radians(eta - arrowAngle))    
        edge_x.append(pointx)
        edge_x.append(pointx - signx**2 * signy * dx)
        edge_x.append(None)
        edge_y.append(pointy)
        edge_y.append(pointy - signx**2 * signy * dy)
        edge_y.append(None)
    
    
    return edge_x, edge_y
```

```python
for node in G.nodes:
    G.nodes[node]['pos'] = pos[node]

edge_x = []
edge_y = []
for edge in G.edges():
    start = G.nodes[edge[0]]['pos']
    end = G.nodes[edge[1]]['pos']
    edge_x, edge_y = addEdge(start, end, edge_x, edge_y, .8, 'end', 25, 25, 10)
    
edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=.6, color='#888'), hoverinfo='none', mode='lines')

node_x = []
node_y = []
for node in G.nodes():
    x, y = G.nodes[node]['pos']
    node_x.append(x)
    node_y.append(y)

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlGnBu',
        reversescale=True,
        color=[],
        size=10,
        colorbar=dict(
            thickness=15,
            title='Aantal connecties',
            xanchor='left',
            titleside='right'
        ),
        line_width=2))


node_adjacencies = []
node_text = []
for node, adjacencies in zip(G.nodes, G.adjacency()):
    node_adjacencies.append(len(adjacencies[1]))
    node_text.append(node+' voor '+str(len(adjacencies[1]))+' artiest(en)')
    
node_trace.marker.color = node_adjacencies
node_trace.text = node_text

fig = go.Figure(data=[edge_trace, node_trace],
             layout=go.Layout(
                title='<b>Nederlandse schrijvers en producenten, en artiesten voor wie ze aan minstens één lied hebben gewerkt</b>',
                titlefont_size=16,
                showlegend=False,
                hovermode='closest',
                margin=dict(b=20,l=5,r=5,t=40),
                xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                )
fig.show()
```

## En samenwerken dan?

Samenwerken is heel logisch in het maken van muziek, dus op vergelijkbare wijze kunnen we een netwerk maken waarin we zien welke Nederlander met een andere Nederlander aan tenminste één liedje samen heeft gewerkt, of dat nou als schrijver, producent of artiest was.

```python
query = """SELECT DISTINCT ?collab1Label ?collab2Label
WHERE 
{
  ?collab1 wdt:P27 wd:Q29999. #allebei NL
  ?collab2 wdt:P27 wd:Q29999.
  ?song (wdt:P86|wdt:P676|wdt:P162|wdt:P175) ?collab1. #schrijver/producent/uitvoerend artiest
  ?song (wdt:P86|wdt:P676|wdt:P162|wdt:P175) ?collab2.
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  FILTER(?collab1 != ?collab2) # niet samenwerken met jezelf
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}
"""

df = sparql_dataframe.get(endpoint_url, query, post=True)
df
```

# Geslachtsverdeling in de muziekindustrie

Met alle informatie in Wikidata is het ook mogelijk om van iedereen die ten minste één lied geschreven/geproduceerd/uitgevoerd heeft dat in Wikidata staat, het geslacht te zien, door middel van Wikidata-property `P21`. Eerst query'en we weer om lijstjes met iedereens geslacht(sidentiteit) op te halen.

```python
# Drie queries om de schrijvers, producenten en artiesten op te vragen met bijbehorende gender.
query_writers = """SELECT DISTINCT ?writerLabel ?genderLabel
WHERE 
{
  ?writer wdt:P27 wd:Q29999. #writer is een Nederlander
  ?writer wdt:P21 ?gender. #gender van de writer
  ?song (wdt:P86|wdt:P676) ?writer. #liedje als path
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""

query_producers = """SELECT DISTINCT ?producerLabel ?genderLabel
WHERE 
{
  ?producer wdt:P27 wd:Q29999. #producer is een Nederlander
  ?producer wdt:P21 ?gender. #gender van de producer
  ?song wdt:P162 ?producer. #liedje als path
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""

query_artiesten = """SELECT DISTINCT ?artistLabel ?genderLabel
WHERE 
{
  ?artist wdt:P27 wd:Q29999. #artiest is een Nederlander
  ?artist wdt:P21 ?gender. #gender van de artiest
  ?song wdt:P175 ?artist. #liedje als path
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""

# Hier worden dataframes gemaakt met de resultaten van de queries.
df_writers = sparql_dataframe.get(endpoint_url, query_writers, post=True)
df_producers = sparql_dataframe.get(endpoint_url, query_producers, post=True)
df_artiesten = sparql_dataframe.get(endpoint_url, query_artiesten, post=True)

# Het weergeven van de dataframes/tabellen.
display(df_writers, df_producers, df_artiesten)
```

# Veel voorkomende beroepscombinaties

Hoe een persoon beschreven staat op een song betekent niet dat dat het enige beroep is dat deze persoon heeft. Het is vaak zo dat mensen binnen de muziek industrie meerdere beroepen invullen. Zo kan er gedacht worden aan schrijvers die ook zingen of producers die ook dj'en.

```python
query_max = """SELECT DISTINCT ?producerLabel ?pberoepLabel
WHERE 
{
  ?producer wdt:P27 wd:Q29999. #producer is een Nederlander
  ?song wdt:P162 ?producer. #liedje als path
  ?producer wdt:P106 ?pberoep.
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""

query_max2 = """SELECT DISTINCT ?writerLabel ?wberoepLabel
WHERE 
{
  ?writer wdt:P27 wd:Q29999.
  ?song (wdt:P86|wdt:P676) ?writer. #releasedatum
  ?writer wdt:P106 ?wberoep.
  {
  ?song wdt:P31 wd:Q7366. #alleen songs
  }
  UNION {
  ?song wdt:P31 wd:Q134556 #en singles
  }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}"""


df_max = sparql_dataframe.get(endpoint_url, query_max, post=True)
df_max2 = sparql_dataframe.get(endpoint_url, query_max2, post=True)
df_max = df_max.sort_values('producerLabel')
df_max2 = df_max2.sort_values('writerLabel')
display(df_max, df_max2)
```

```python
prods = set()
prods_that_are = set()
for prod in df_max.index:
    prods.add(df_max['producerLabel'][prod])
    if df_max['pberoepLabel'][prod] == 'diskjockey' or df_max['pberoepLabel'][prod] == 'live-dj':
        prods_that_are.add(df_max['producerLabel'][prod])

writers = set()
writers_that_are = set()
for wri in df_max2.index:
    writers.add(df_max2['writerLabel'][wri])
    if df_max2['wberoepLabel'][wri] == 'zanger' or df_max2['wberoepLabel'][wri] == 'singer-songwriter':
        writers_that_are.add(df_max2['writerLabel'][wri])


import matplotlib.pyplot as plt
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
label = 'Percentage writers dat ook zanger is'
labels = 'writers', 'sing-songwriter'
sizes = [len(writers)-len(writers_that_are), len(writers_that_are)]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title(label)
plt.show()
label = 'Percentage producers dat ook dj is'
labels = 'producer', 'dj'
sizes = [len(prods)-len(prods_that_are), len(prods_that_are)]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title(label)
plt.show()
```

Door Wikidata te query'en hebben een klein beetje inzicht gekregen in de Nederlandse muziekindustrie, wat best veel verrassend was. Helaas waren er ook heel veel extra dingen die we wilden doen, maar niet konden door beperkingen of ontbrekende data in Wikidata. Voorbeelden hiervan:

* Hitlijstdata (property bestaat, was echt bijna nooit ingevuld)

* Van sommige mensen, zoals Marco Borsato ook geen geslacht ingevuld.
* Groepsleden van groepen niet altijd ingevuld, dus die konden we niet altijd wegfilteren of apart visualiseren.
* Vaak hebben alleen populaire liedje een Wikipedia pagina, dus staan de niet-populaire liedjes ook niet in Wikidata.
* Inconsistenties: Soms is Enschede (gemeente) en soms Enschede (stad) ingevuld, of bij een paar mensen alleen de provincie (Noord-Brabant)
* Stel iets is niet ingevoerd: valt weg uit hele _conjunctive query_.
