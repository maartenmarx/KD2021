```python
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://query.wikidata.org/sparql"

query = """select distinct ?countryLabel
where {
  ?division wdt:P2094 wd:Q31930761 .
  ?division rdfs:label ?label_en filter (lang(?label_en) = "en").
  ?division wdt:P17 ?country .
  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "nl"}.
}
order by asc(str(?countryLabel))"""


def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    # TODO adjust user agent; see https://w.wiki/CX6
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()


results = get_results(endpoint_url, query)


```

Hieronder worden de landen die een voetbaldivisie hebben uit wikidata gehaald:

```python
url = 'https://query.wikidata.org/sparql'
query = """select distinct ?countryLabel ?country
                                    where {
                                    ?division wdt:P2094 wd:Q31930761 .
                                    ?division rdfs:label ?label_en filter (lang(?label_en) = "nl").
                                    ?division wdt:P17 ?country .
                                        
                                    SERVICE wikibase:label { bd:serviceParam wikibase:language "nl"}.
                                    }
                                    order by asc(str(?countryLabel))"""
r = requests.get(url, params = {'format': 'json', 'query': query})
countrydata = r.json()

countries = []
for item in countrydata['results']['bindings']:
    countries.append(OrderedDict({
        'country': item['countryLabel']['value'],
        'countrycode': item['country']['value']
    }))
    

countrydf = pd.DataFrame(countries)
countrydf.set_index('country', inplace=True)
display(countrydf)

display(countrydf.loc["Nederland"])

print("Nederland heeft dus de entiteit wd:Q55")
```

Natuurlijk laten wij even aan de hand van het Nederlandse voetbal zien welke divisies er zijn:

```python
url = 'https://query.wikidata.org/sparql'
query = """
select ?divisionLabel ?division
                                    where {
                                        ?division wdt:P2094 wd:Q31930761 .
                                        ?division rdfs:label ?label_en filter (lang(?label_en) = "nl").
                                        ?division wdt:P17 wd:Q55 .
                                        ?division wdt:P31 wd:Q15991303.
                                        SERVICE wikibase:label { bd:serviceParam wikibase:language "nl"}.
                                    }"""
r = requests.get(url, params = {'format': 'json', 'query': query})
divisiondata = r.json()

divisions = []
for item in divisiondata['results']['bindings']:
    divisions.append(OrderedDict({
        'division': item['divisionLabel']['value'],
        'divisioncode': item['division']['value']
        
        
    }))
    

divisiondf = pd.DataFrame(divisions)
divisiondf.set_index('division', inplace=True)
display(divisiondf)
print("De eredivisie heeft dus entiteit wd:Q167541")
```

Vanuit die divisie, in dit geval dus de Eredivisie, laten we de verschillende clubs zien die daar in spelen:

```python
url = 'https://query.wikidata.org/sparql'
query = """
SELECT ?clubLabel ?club
                                    WHERE {
                                        ?club wdt:P31 wd:Q476028 .
                                        ?club wdt:P118 wd:Q167541 .
                                        SERVICE wikibase:label{
                                            bd:serviceParam wikibase:language "nl" .
                                        }
                                    }"""
r = requests.get(url, params = {'format': 'json', 'query': query})
data = r.json()

clubs = []
for item in data['results']['bindings']:
    clubs.append(OrderedDict({
        'club': item['clubLabel']['value'],
        'clubcode': item['club']['value']
        
    }))
    

df = pd.DataFrame(clubs)
df.set_index('club', inplace=True)
display(df)
display(df.loc["AFC Ajax"])

print("Ajax heeft dus entiteit wd:Q81888")
```
