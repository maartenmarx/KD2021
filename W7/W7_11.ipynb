```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import folium
from urllib.request import urlopen
from PIL import Image 

from msticpy.nbtools import *
from msticpy.sectools import *
from datetime import datetime
```

Uitleg SPARQL

De informatie die wij in dit werk gebruiken halen wij uit Wikidata door middel van SPARQL. Voor extra informatie over het gebruik van SPARQL in combinatie met Wikidata, zie https://towardsdatascience.com/where-do-mayors-come-from-querying-wikidata-with-python-and-sparql-91f3c0af22e2.

Door middel van de onderstaande query vragen wij van Wikidata alle Nederlandse schilders op. Van deze Nederlandsche schilders halen wij vervolgens ook bekende werken met hun huidige locatie op, zodat deze later op de kaart weergegeven kunnen worden. Steden en landen als locatie worden uit de lijst gefilterd. Ten slotte wordt er nog overige informatie opgehaald, waarvan sommige informatie optioneel, zoals geboortedatum, geslacht en plaatjes van de kunstwerken.


```python
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")


sparql.setQuery("""

SELECT ?painter ?painterLabel
       ?notableWork ?notableWorkLabel ?creationdateLabel
       ?location ?locationLabel ?coordinates
       ?image ?image2 ?sexLabel ?birthdateLabel ?deathdateLabel
        
WHERE {
  ?painter wdt:P106 wd:Q1028181 .
  ?painter wdt:P172 wd:Q200569 .
  ?painter wdt:P800 ?notableWork .
  ?notableWork wdt:P276 ?location .

  ?painter wdt:P21 ?sex .
  ?painter wdt:P569 ?birthdate .
  OPTIONAL {?painter wdt:P570 ?deathdate}  

  FILTER NOT EXISTS {
     ?location wdt:P31 wd:Q515 
   }
    FILTER NOT EXISTS {
     ?location wdt:P31 wd:Q6256
   }
  
  ?location wdt:P625 ?coordinates
  OPTIONAL {?notableWork wdt:P571 ?creationdate }
  OPTIONAL {?notableWork wdt:P18 ?image }
  OPTIONAL {?painter wdt:P18 ?image2 }

  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
ORDER BY ?notableWork ?notableWorkLabel
""")

sparql.setReturnFormat(JSON)
results = sparql.query().convert()

results_df = pd.json_normalize(results['results']['bindings'])
results_df
```

In de codecel hieronder wordt gebruik gemaakt van een ipywidget. Deze maakt het mogelijk python input interactief te maken. Door een lijst van Nederlandse schilders uit de query hierboven te pakken gaat deze in zijn werking.


```python
"""Dropdown menu kiezen schilder"""

# Maak een set met alle schilders in de dataset
schilders = set()
for index, row in results_df.iterrows():
  df = results_df.iloc[index]
  schilders.add(df['painterLabel.value'])

# Voeg de schilders toe aan de dropdown
schilder = list(schilders)[0]
ws = widgets.Dropdown(
    options=list(schilders),
    value=list(schilders)[0],
    description='Schilder:',
)

# Controleer op interacties met de dropdown
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        global schilder 
        schilder= change['new']

ws.observe(on_change)

display(ws)
```

De codecel hieronder maakt het vervolgens mogelijk om een klein beetje data bij de gekozen schilder in te zien.

```python
# Data schilder
for index, row in results_df.iterrows():
  df = results_df.iloc[index]
  if df['painterLabel.value'] == schilder:
    break

# Print info
w = df['image2.value'] + '?width=300'
print('Naam: ' + df['painterLabel.value'])
print('Gender: ' + df['sexLabel.value'])
print('Geboortedatum: ' + df['birthdateLabel.value'][0:10])
print('Sterfdatum: ' + df['deathdateLabel.value'][0:10])
Image.open((urlopen(w)))
```

Hieronder is een tijdlijn met alle werken van de gekozen schilder. Per locatie is er een ander kleurtje, zo kan snel worden gezien hoeveel schilderijen van de gekozen schilder er op welke locatie zijn. Als de muis op een punt wordt gehouden kan er extra informatie gezien worden.

```python
selectedWorks_df = results_df.loc[results_df['painterLabel.value'] == schilder]

selectedWorks_df = selectedWorks_df[['painterLabel.value',
                                     'notableWork.value',
                                     'notableWorkLabel.value',
                                     'locationLabel.value',
                                     'location.value',
                                     'image.value',
                                     'creationdateLabel.value']]

selectedWorks_df['creationdateLabel.value'] = selectedWorks_df['creationdateLabel.value'].apply(lambda x: datetime.strptime(x[:4], '%Y'))

selectedWorks_df = selectedWorks_df.rename(columns={'painterLabel.value':'Painter',
                                                    'notableWork.value':'Link',
                                                    'locationLabel.value': 'Location',
                                                    'creationdateLabel.value':'Creation',
                                                    'notableWorkLabel.value':'Name'})

selectedWorks_df = selectedWorks_df.drop_duplicates(subset=['Name'])

title = str('Schilderijen door ' + schilder)
nbdisplay.display_timeline(selectedWorks_df, 
                           range_tool=False, 
                           title=title, 
                           source_columns=['Painter', 'Link', 'Name', 'Location'], 
                           time_column='Creation',
                           group_by='Location',
                           legend='left')
```

De codecel hieronder maakt het mogelijk om een specifiek werk in meer detail te bekijken.


```python
"""Dropdown menu kiezen notableworks"""

# Maak een set met alle werken van een schilder in de dataset
works = set()
for index, row in results_df.iterrows():
  df = results_df.iloc[index]
  if df['painterLabel.value'] == schilder:
    works.add(df['notableWorkLabel.value'])

# Voeg de schilders toe aan de dropdown
work = list(works)[0]
ww = widgets.Dropdown(
    options=list(works),
    value=list(works)[0],
    description='Schilderij:',
)

# Controleer op interacties met de dropdown
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        global work
        work = change['new']

ww.observe(on_change)

display(ww)
```

```python
# Info gekozen schilderij inladen
df = results_df.loc[results_df['notableWorkLabel.value'] == work]

"""Info kunstwerk"""

print('Titel: ' + work)
print('Creatiedatum: ' + str(df['creationdateLabel.value']))

# Display the image
for url in df['image.value']:
  w = url
w += '?width=1024'

Image.open((urlopen(w)))
```

```python
In de codecel hieronder word gebruik gemaakt van de Python module folium. Deze maakt het mogelijk een interactieve kaart te maken binnen Jupyter Notebook. Met behulp van cöordinaten kan vervolgens de locatie van een werk aangegeven worden met een 'marker'.
```

```python
# Locatie gekozen werk wordt ingeladen op kaart
df = results_df.loc[results_df['notableWorkLabel.value'] == work]
for v in df['coordinates.value']:
  w = v
for v in df['locationLabel.value']:
  l = v

# De cöordinaten leesbaar maken voor folium
cordstr = w[6:-1].split()

# Coordinaten wisselen
cordstr.reverse()

cords = []
for item in cordstr:
  cords.append(float(item))

# De kaart inladen met de cöordinaten van het werk als middelpunt
m = folium.Map(location=cords)

# Voeg de marker toe met de naam van de locatie van het werk.
folium.Marker(
  location=cords,
  popup=l,
).add_to(m)

m
```
