# Het probleem
## Hoe presteren Nederlanders op de Olympische spelen?
We kregen een beetje de kriebels van het zelf niet meer mogen sporten door corona en dus waren we benieuwd hoe goed wij als land nou eigenlijk zijn in sport. Daarom kijken wij voor deze opdracht naar Nederlanders die ooit mee hebben gedaan aan een onderdeel van de Olympische spelen en goed hoe ze daar hebben gepresteerd. We kijken zowel naar de Zomerspelen als de Winterspelen. 

Om hier een antwoord op te krijgen, hebben we op wikidata alle participaties aan onderdelen van Olympische spelen opgezocht, waarin de participanten Nederlands zijn.

Per participatie halen we de volgende kenmerken op door middel van de SPARQL query hieronder:
- De participant
- Geboorteplaats en coördinaten van geboorteplaats van de participant
- Het soort Olympische Spelen (Zomer- of Winterspelen)
- De specifieke spelen waaraan is meegedaan (b.v '2000 Summer Olympics')
- Het specifieke sportonderdeel waaraan de participant heeft meegedaan (b.v 'cycling at the 1928 Summer Olympics – men's tandem')
- Het jaartal waarin de Spelen van de participatie plaatsvond
- Eventueel gewonnen medaille

```python
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://query.wikidata.org/sparql"

query = """SELECT DISTINCT  ?player ?playerLabel ?spelen ?spelenLabel ?spelenSpecifiek ?spelenSpecifiekLabel ?spelenJaar ?spelenOnderdeel ?spelenOnderdeelLabel  ?sport ?sportLabel ?medalLabel ?geboorteplaats ?geboorteplaatsLabel ?geboorteplaatsCoordinaten

WHERE{
  ?player wdt:P27 wd:Q29999;                                                          #?player -- country of citizenship -> Kingdom of the Netherlands
          wdt:P1344 ?spelenOnderdeel;                                                 #        -- participant in ---------> ?spelenOnderdeel
          wdt:P19 ?geboorteplaats.                                                    #        -- place of birth ---------> ?geboorteplaats
  ?geboorteplaats   wdt:P625 ?geboorteplaatsCoordinaten.                              #?geboorteplaats -- coordinate location -> ?geboorteplaatsCoordinaten
    
  ?spelen  wdt:P279 wd:Q5389. FILTER(?spelen = wd:Q159821 || ?spelen = wd:Q82414)     #?spelen -- subclass of --> Olympic Games  FILTER(?spelen = Summer Olympic Games || ?spelen = Winter Olympic Games)
  ?spelen wdt:P527 ?spelenSpecifiek.                                                  #?spelen -- has part -----> ?spelenSpecifiek
  
  ?spelenOnderdeel wdt:P31*/wdt:P361 wd:Q26132862;                                    #?spelenOnderdeel --instance of*/part of--> Olympic sports discipline event
                   wdt:P361* ?spelenSpecifiek;                                        #                 -- part of -------------> ?spelenSpecifiek
                   ?afgespeeldIn  ?spelenDatum;                                       #                 -- ?afgespeeldIn -------> ?spelenDatum
                   wdt:P641 ?sport                                                    #                 -- sport ---------------> ?sport
           
  FILTER(?afgespeeldIn = wdt:P585 || ?afgespeeldIn = wdt:P580).                       #FILTER(?afgespeeldIn = point in time || ?afgespeeldIn = start time)
                    
  BIND(YEAR(?spelenDatum) as ?spelenJaar).                                            
 
  OPTIONAL{ # medaille via property 'participant in' indien bestaand
    ?player p:P1344 ?participateStatement.                                            #?player -- participant in ---------------> ?participateStatement
    ?participateStatement ps:P1344 ?spel. FILTER(?spel = ?spelenOnderdeel).           #?participateStatement --participant in --> ?spel
    ?participateStatement pq:P166 ?medal.                                             #                      -- award received -> ?medal
  }
 
  SERVICE wikibase:label { bd:serviceParam wikibase:language "nl, en". }
}"""
```

## Resultaten ophalen en omzetten naar een dataframe

```python
from pandas import json_normalize

def get_results(endpoint_url, query):
    sparql = SPARQLWrapper(endpoint_url)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)

    result = sparql.query().convert()
    return json_normalize(result["results"]["bindings"])


resultsDF_original = get_results(endpoint_url, query)
```

```python
resultsDF = resultsDF_original[[column for column in resultsDF_original.columns if ".value" in column]]
resultsDF = resultsDF.rename(columns = lambda col: col.replace(".value", ""))
resultsDF.head()
```

## Beter? Beter.
Maar oh wacht, voor de library die we zometeen gaan gebruiken, Folium, moeten de coördinaten omgedraaid en zonder text en haken zijn. Dus niet zoals hieronder.

```python
resultsDF['geboorteplaatsCoordinaten'][0]
```

Ook aanpassen dus!

```python
# Coordinaten draaien om ze bruikbaar te maken voor Folium zometeen
def prepareCoordinates(coordSeries):
    for i in range(len(coordSeries)):
        coordSeries[i] = coordSeries[i].strip('Point()')
        coordSeries[i] = [coord for coord in coordSeries[i].split(' ')]
        coordSeries[i][0] , coordSeries[i][1] = coordSeries[i][1] , coordSeries[i][0]
        coordSeries[i] = (str(coordSeries[i][0]), str(coordSeries[i][1]))

prepareCoordinates(resultsDF['geboorteplaatsCoordinaten'])
```

Oké, en verder?

Nou, we willen op een kaart alle personen, de sport die ze beoefenen, hoeveel participaties ze hebben en hoeveel medailles ze hebben gewonnnen. Daarvoor kunnen we onze dataframe van eerder groeperen op personen (met bijhorende coördinaten), sport, participaties en medailles.

```python
participatiesPerPersoon = resultsDF.groupby(['playerLabel', 'player', 'geboorteplaatsLabel', 'geboorteplaatsCoordinaten', 'sport', 'sportLabel']).count()[['spelen', 'medalLabel']]
participatiesPerPersoon.sort_values(by='medalLabel', ascending=False).head()
```

Vervolgens kunnen we met alleen de informatie uit deze dataframe en de library Folium een mooie interactieve kaart maken waarop alle personen uit de dataframe worden afgebeeld, met wat extra informatie.

```python
import folium 
from folium.plugins import MarkerCluster

mapPersons = folium.Map(location=[52.370216, 4.895168], control_scale = True, zoom_start=10)
marker_clusterPersons = MarkerCluster().add_to(mapPersons)

for index, value in participatiesPerPersoon.iterrows():
    playerLabel, player, geboorteplaatsLabel, geboorteplaatsCoordinaten, sport, sportLabel, aantalParticipaties, aantalMedailles = index[0], index[1], index[2], index[3], index[4], index[5], value[0], value[1]

    popupPersons = folium.Popup(f"""   Speler: <br>
                                <a href={player}>{playerLabel}</a> 
                                <br><br>
                                Sport:<br> 
                                <a href={sport}>{sportLabel}</a>  
                                <br><br>
                                Participaties:<br> 
                                {aantalParticipaties} 
                                <br><br>
                                Aantal gewonnen Olympische medailles:<br>
                                {aantalMedailles}""", 
                                max_width = 200)
    folium.Marker(geboorteplaatsCoordinaten,popup= popupPersons,tooltip=playerLabel).add_to(marker_clusterPersons)
```

En dan afbeelden in een mooie, interactieve grafiek via plotly.

```python
import plotly.graph_objs as go
participatieMedailles = go.Figure(
    data=[
        go.Bar(name='Aantal participaties', x=perJaar.index, y=perJaar['player'], width=2),
        go.Bar(name='Aantal gewonnen medailles', x=perJaar.index, y=perJaar['medalLabel'], width=1.5)
    ]
)
participatieMedailles.update_layout(
    autosize=False,
    width=1000,
    height=550,
    margin=dict(
        l=10,
        r=10,
        b=30,
        t=50,
    ),
    title_text='Aantal Nederlandse participaties en gewonnen medailles door de jaren heen',
    yaxis=dict(title_text='Aantal', range=[0, 80]),
    xaxis=dict(title_text='Jaartal', tickvals=perJaar.index, tickangle = -70)
)
```

Verder leek het ons leuk om te laten zien in welke  jaren een zelf gekozen aantal medailles gewonnen zijn, per sport. Hiervoor moeten we onze dataframe dus groeperen op jaar en sport.

```python
sportPerJaar = resultsDF.groupby(['spelenJaar', 'sportLabel']).count()[['player', 'medalLabel']]
sportPerJaar.head()
```

Vervolgens gaan we widgets gebruiken! 

Gebruiksaanwijzing:

```python
import ipywidgets as widgets
from ipywidgets import HBox, Label, Layout

aantalMedailles = widgets.IntSlider(
    value=5,
    min=sportPerJaar['medalLabel'].min(),
    max=sportPerJaar['medalLabel'].max(),
    description='Minimum aantal gewonnen medailles:',
    layout=Layout(
        width='50%',
    ),
    style={'description_width': 'initial'}
)
display(aantalMedailles)
```

```python
from collections import defaultdict
q = [(i[0], i[1]) for i in sportPerJaar[sportPerJaar['medalLabel'] >= aantalMedailles.value].index]

sportMedailles = defaultdict(list)
for result in q:
    sportMedailles[result[1]].append(result[0])

for i in sportMedailles.items():
    if aantalMedailles.value == 1 and len(i[1]) > 1:
        print(f"Bij {i[0]} is in de jaren {i[1]} minimaal {aantalMedailles.value}  medaille gewonnen")
    elif len(i[1]) > 1:
        print(f"Bij {i[0]} zijn in de jaren {i[1]} minimaal {aantalMedailles.value}  medaille(s) gewonnen")
    else:
        print(f"Bij {i[0]} zijn in het jaar {i[1][0]} minimaal {aantalMedailles.value}  medaille(s) gewonnen")
```

# Reflectie

Dit is de eerste keer dat we voor een project geheel met wikidata als informatiebron werken. Hierdoor zijn we erachter gekomen dat, hoe prachtig het idee van wikidata ook is, het ook erg onpraktisch kan zijn bij het gebruik van wikidata-data (ha) in een daadwerkelijk project. Omdat iedereen op elk moment pagina’s kan aanpassen en omdat wikidata grotendeels afhankelijk is van individuele mensen die hopelijk op een goede manier informatie invoeren, is wikidata als database eigenlijk best onbetrouwbaar. Een duidelijk voorbeeld hiervan, waar wij tegenaan liepen, is dat de ‘instance of’ statement van alle (x jaar) Winter Olympics was veranderd, waardoor onze query ineens geen winterspelen meer oppakte, terwijl het dat eerder wel deed. Ook mist veel informatie op wikidata, of wordt er op inconsequente manieren omgegaan met het invoeren van informatie. Veel mensen zijn bijvoorbeeld ‘participant of’ een ‘(x jaar) Summer Olympics’, terwijl andere mensen ‘participant of’ een specifiek onderdeel van een Spelen. In onze query kijken we alleen naar mensen bij wie wordt aangegeven dat ze direct hebben meegedaan aan een specifiek onderdeel van een Spelen, zodat ook duidelijk is welke sport ze hebben beoefend en waarvoor ze eventueel een medaille hebben gewonnen. Dit betekent echter wel dat veel participanten niet zijn meegenomen. Al met al hebben we veel geleerd door te werken met wikidata en beseffen we ook dat er nog veel te leren valt.
