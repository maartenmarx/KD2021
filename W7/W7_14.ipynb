# Wikidata tutorial  

Voor deze tutorial hebben we gebruik gemaakt van films op wikidata die zijn gemaakt door een Nederlands bedrijf of Nederlandse bedrijven. Van deze films wilden wij informatie verzamelen. Het probleem dat we tijdens deze tutorial gaan behandelen is dat we via wikidata geen ratings opvragen van de films. Deze ratings wilden we natuurlijk wel krijgen, omdat dat belangrijke en leuke data is om te visualiseren en te gebruiken in deze tutorial. De ratings moesten we dus verkrijgen via een andere weg. Eerst gingen we inventariseren waar deze ratings van de films beschikbaar waren. We stuitten al snel op de database van IMDB, die een scala aan ratings van films ter beschikking heeft. Vervolgens gingen we via wikidata zoeken naar een mogelijke link tussen de films en de ratings die beschikbaar waren bij IMDB. We ontdekten tijdens de uitwerking van de tutorial dat we via wikidata wél de IMDB ID's konden opvragen van de films. Belangrijk hierbij is dat veel films een IMDB ID hebben maar helaas niet alle films hebben deze ID's. Wij hebben er dus voor gekozen om alleen verder te gaan met films die deze ID's wel hebben. Door middel van deze ID' s is het mogelijk om veel informatie per film op te vragen bij de IMDB database. Hier zaten ook de ratings tussen die we wilden gebruiken voor deze tutorial. 


Voordat we beginnen gaan we eerst alle imports noteren. In de comments staan de installs die vereist zijn om de tutorial te kunnen runnen.

```python
# %pip install html
# %pip install lxml
# %pip install plotly
# %pip install folium
# %pip install shapely
# %pip install SPARQLWrapper
# %pip install collections

import html
import folium
import requests
import shapely.wkt
import numpy as np
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from lxml import html as HTML
from ipywidgets import interact
from collections import OrderedDict
from folium.plugins import MarkerCluster
from SPARQLWrapper import SPARQLWrapper, JSON
```

## Wikidata in Nederland gemaakte films

Via wikidata worden met SPARQL queries alle films opgevraagd die een Nederlandse oorsprong hebben. In deze dataset hebben we alleen de films gepakt die de subclass filming location bevatten. Een film kan meerdere film locaties hebben. Deze dataset wordt in de cel hieronder opgevraagd via een query die gebruik maakt van de wikidata query service. De data wordt hieronder weergegeven in een tabel.

```python
url = 'https://query.wikidata.org/sparql'
query = '''
SELECT DISTINCT ?movieLabel ?languageLabel ?genreLabel ?imdb ?coordinates

WHERE {
  ?movie wdt:P495 wd:Q55.
  ?movie wdt:P31 wd:Q11424.
  ?movie wdt:P364 ?language.
  ?movie wdt:P136 ?genre.
  OPTIONAL { ?movie wdt:P345 ?imdb. }

  SERVICE wikibase:label { bd:serviceParam wikibase:language "nl". }
  }
'''
r = requests.get(url, params = {'format': 'json', 'query': query})
data = r.json()

movies = []
for item in data['results']['bindings']:
    movie_data = OrderedDict({
        'movie': item['movieLabel']['value'],
        'language': item['languageLabel']['value'],
        'genre': item['genreLabel']['value']
    })
    if 'imdb' in item:
        movie_data.update({'imdb_id': item['imdb']['value']})
    if 'coordinates' in item:
        movie_data.update({'coordinates': item['coordinates']['value']})

    movies.append(movie_data)

df = pd.DataFrame(movies)
df
```

## Films met hun IMDB Ratings

Hieronder volgt een tabel waarin de films staan. Hierin staat de language die aangeeft in welke taal de film is opgenomen, het genre van de film, de IMDB ID van de film en de coördinaten van de locatie waar de films zijn opgenomen. Sommige films hebben meerdere coördinaten en zullen dus ook meerdere locaties hebben op de kaart die later volgt. In de tabel daaronder worden vervolgens ook de IMDB ratings toegevoegd van de films. De genres van de films en de ratings van de films zullen ook een rol spelen in de staafdiagram die volgt waarin de gemiddelde ratings per genre worden weergegeven.

```python
def lookup(url, ratings):
    r = requests.get(url)
    page = HTML.fromstring(r.content)

    # zoekt ratings
    movies = page.xpath('//div[@class="lister-item-content"]')
    for movie in movies:
        imdb_id = str(movie.xpath('./h3/a/@href'))[9:-3]
        rating = movie.xpath('./div/div/strong/text()')
        if rating:
            ratings[imdb_id] = rating[0]
    
    # volgende pagina
    next_page = page.xpath('//a[@class="lister-page-next next-page"]/@href')
    if next_page and 'start=1501' not in str(next_page):
        lookup('https://www.imdb.com' + next_page[0], ratings)
        
    return pd.DataFrame.from_dict(ratings, orient='index', columns=['rating'])

ratings_df = lookup('https://www.imdb.com/search/title/?title_type=feature,tv_movie,documentary&countries=nl&count=250', dict())
```

```python
imdb_df = pd.DataFrame(movies).set_index('imdb_id')
df2 = imdb_df.join(ratings_df).dropna()
df2['rating'] = pd.to_numeric(df2["rating"], downcast="float")

df2
```

## Filmlocaties van in Nederland gemaakte films

Met de opgevraagde wikidata dataset met alle in Nederland gemaakte films zijn de filmlocaties bij veel films bekend. Deze filmlocaties hebben allemaal een eigen coördinaat, en deze kunnen ook worden weergegeven in de wikidata query service. Er ontstaan echter problemen wanneer deze coördinaten worden gebruikt om een kaart te maken de Folium library van python. De coördinaten staan namelijk in WKT format en dit format is onbruikbaar wanneer gebruikt met Folium. Om dit probleem op te lossen kun je gebruik maken van Shapely. Hiermee kunnen WKT coördinaten worden omgezet naar een bruikbaar format. Wanneer dit is gedaan kan er een kaart worden gemaakt. Deze interactieve kaart bevat alle filmlocaties. Het is mogelijk in te zoomen en te klikken op een filmlocatie. Wanneer dit wordt gedaan komt er een pop-up met informatie naar voren. In deze pop-up staat de naam van de film die is opgenomen op die locatie en de naam van de locatie. Belangrijk om te weten bij deze kaart is dat in de wikidata dataset niet elke film de subclass filming location bevat. Dit betekent dat op de kaart alleen filmlocaties zijn weergegeven van films die deze subclass wel bevatten. Desalniettemin is toch de keuze gemaakt deze kaart weer te geven. Ook is het belangrijk om te weten dat sommige films meerdere filmlocaties hebben, dus voor één film kunnen meerdere locaties op de kaart zijn weergegeven.  


```python
wc_sparql = """
SELECT DISTINCT ?movieLabel ?languageLabel ?locationLabel ?coordinates

#defaultView:Map 

WHERE {
  ?movie wdt:P495 wd:Q55.
  ?movie wdt:P31 wd:Q11424.
  ?movie wdt:P364 ?language.
  ?movie wdt:P915 ?location.
  ?location wdt:P625 ?coordinates.

  SERVICE wikibase:label {
     bd:serviceParam wikibase:language "nl" .
   }
  }

"""

sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
sparql.setQuery(wc_sparql)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

movie = {}
for result in results["results"]["bindings"]:
    coords = shapely.wkt.loads(result["coordinates"]["value"]).coords[0]

    movielocation = result["locationLabel"]["value"]
    moviename = result["movieLabel"]["value"]
    movie[movielocation] = {
            "lat_lng": coords,
            "movie": moviename,
            "country": movielocation,
        }

moviemap = folium.Map(
    location=[20,0],
    zoom_start=2,
    tiles='Stamen Toner',
    attr='''<a id="home-link" target="_top" href="../">Map tiles</a> by 
    <a target="_top" href="http://stamen.com">Stamen Design</a>, 
    under <a target="_top" href="http://creativecommons.org/licenses/by/3.0">CC BY 3.0</a>. 
    Data by <a target="_top" href="http://openstreetmap.org">OpenStreetMap</a>, 
    under <a target="_top" href="http://creativecommons.org/licenses/by-sa/3.0">CC BY SA</a>.
    | Locations powered by <a href="https://query.wikidata.org/">Wikidata</a>.'''
)

clusters = {}
for movie_id in movie:

    mov = movie[movie_id]
    if mov["country"] not in clusters:
        clusters[mov["country"]] = MarkerCluster().add_to(moviemap)
    folium.Marker(
        [mov["lat_lng"][1], mov["lat_lng"][0]],
        popup='Movie: {} - FilmingLocation: {} '.format(
            html.escape(mov["movie"]), 
            html.escape(mov["country"])),
        icon=folium.Icon(icon='fas fa-film', prefix='fa')
        ).add_to(clusters[mov["country"]])

moviemap
```
