### 1.2 Het aanmaken van de familie relaties

Er wordt eerst een networkx netwerk aangemaakt met de functie nx.DiGraph()  
Vervolgens worden alle mannen en vrouwen in de familie van Thomas hieraan toegevoegd als knopen  
Daarna krijgt elke node de waarde True of False gebaseerd op of die node een man of een vrouw is en krijgt elke knoop een bijbehorende kleur  
Als laatste worden er lijnen aangemaakt tussen de knopen die een ouder-kind relatie hebben

```python
%matplotlib inline

# Familienetwerk Thomas
T = nx.DiGraph()

# De mannen en vrouwen in het netwerk
Man,Vrouw = {'Albert','Ton','Rene','Bartel','Dick','Erik','Paul_G',
            'Gijs','Paul_dJ','Thomas','Bart','Arjan','Sil','Sverre',
            'Owen','Fabe'}, {'Tiny','Aafke', 'Marleen','Ellen','Yvonne_dJ','Marianne',
            'Yvonne_D','Laura','Wieneke','Jade','Linde'}

# Alle mensen toevoegen aan het netwerk
T.add_nodes_from(Man)
T.add_nodes_from(Vrouw)

# Nu even de attributen M en V goed zetten
for n in Man:
    T.nodes[n]['M'] = True
    T.nodes[n]['V'] = False

for n in Vrouw:
    T.nodes[n]['V'] = True
    T.nodes[n]['M'] = False

# Voeg kleuren toe aan nodes
kleurenT = []
for node in T.nodes:
    if node in Man:
        kleurenT.append('lightblue')
    elif node in Vrouw:
        kleurenT.append('pink')

# Toevoegen van alle ouder-kind relaties in de familie
T.add_edges_from([('Albert','Rene'), ('Albert','Bartel'), ('Albert','Dick'), ('Albert','Marleen'),
                  ('Tiny','Rene'), ('Tiny','Bartel'), ('Tiny','Dick'), ('Tiny','Marleen'),
                  ('Ton','Ellen'), ('Ton','Yvonne_dJ'), ('Ton','Marianne'), ('Ton','Paul_dJ'),
                  ('Aafke','Ellen'), ('Aafke','Yvonne_dJ'), ('Aafke','Marianne'), ('Aafke','Paul_dJ'),
                  ('Rene','Yvonne_D'), ('Rene','Thomas'), ('Rene','Laura'),
                  ('Ellen','Yvonne_D'), ('Ellen','Thomas'), ('Ellen','Laura'),
                  ('Paul_G','Sil'), ('Paul_G','Sverre'), ('Paul_G','Fabe'),
                  ('Yvonne_dJ','Sil'), ('Yvonne_dJ','Sverre'), ('Yvonne_dJ','Fabe'),
                  ('Gijs','Jade'), ('Gijs','Linde'), ('Gijs','Owen'),
                  ('Marianne','Jade'), ('Marianne','Linde'), ('Marianne','Owen'),
                  ('Marleen','Bart'), ('Marleen','Arjan'), ('Marleen','Wieneke'),
                  ('Erik','Bart'), ('Erik','Arjan'), ('Erik','Wieneke')])
```

### 1.3 Het networkx netwerk omzetten naar RDF
Om het networkx netwerk om te zetten naar RDF moeten alle ouder-kind relaties en alle node attributen uit het networkx netwerk in de RDF database gezet worden.

De RDF maakt gebruik van de "https://Familieboom.org" namespace.

Vervolgens moet je de Namespace binden aan de RDF database graaf in dit geval 'g'.

Nu kan je beginnen met het toevoegen van nodes van de networkx graaf G aan de RDF database graaf g door gebruik te maken van de functie g.add((subject, predicaat, object))

```python
def networkx2RDF(G):
    g = Graph()
    n = Namespace("https://familieboom/naam.org")
    p = Namespace("https://familieboom/ouder.org")
    g.bind('n', n)
    g.bind('p', p)

    # Toevoegen van ouder-kind relaties en naam
    for x,y in G.edges():
        g.add((n[str(x)], n.name, Literal(x)))
        g.add((n[str(y)], n.name, Literal(y)))
        g.add((n[str(x)], p.parent, n[str(y)]))

    # Toevoegen van attributen van de nodes
    nodes = dict(G.nodes(data=True))
    for key,val in nodes.items():
        for k,v in nodes[key].items():
            if v:
                g.add((n[key], n.type, n[k]))

    # Print en return de nieuwe RDF database
    print(g.serialize(format="turtle").decode("utf-8"))
    return g

g = networkx2RDF(T)
```
